# Importing flat files from the web: your turn!

A continuación vamos a importar nuestro primer archivo desde la web. El fichero que importaremos será **winequality-red.csv**. Este fichero contiene datos tabulares con información química respecto a los vinos tintos. Para importar datos desde la web haremos uso de la librería **urllib**.

```python
#Importamos la librería y sus paquetes
from urllib.request import urlretrieve
import pandas as pd

#Asignamos a una variable la url donde importaremos el fichero 
url = 'https://s3.amazonaws.com/assets.datacamp.com/production/course_1606/datasets/winequality-red.csv'

#Salvamos el fichero de forma local 
urlretrieve(url, 'winequality-red.csv')
```

Ahora para comprobar que la descarga se ha producido de forma correcta vamos a cargar los datos y mostrar las primeras observaciones.

```python
#Leemos y  mostramos las primeras observaciones
df = pd.read_csv('winequality-red.csv', sep = ';')
print(df.head())
```

# Performing HTTP requests in Python using urllib

El paquete **Request** se trata de uno de los paquetes más usados de Python. Este paquete entre otras cosas nos permite realizar peticiones a páginas webs y recoger la respuesta.

```python
#Importamos paquetes
from urllib.request import urlopen, Request

#Nos creamos una variable con la url a la cual queremos realizar la petición 
url = "http://www.datacamp.com/teach/documentation" 

#Preparamos la petición a la url
request = Request(url)

#Hacemos la petición y capturamos la respuesta
response = urlopen(request)

#Extraemos la respuesta
html = response.read()

#Mostramos la respuesta 
print(html)

#Cerramos la conexión
response.close()
```

# Performing HTTP requests in Python using requests

Ahora que tenemos claro como se realiza una petición mediante el paquete urllib, vamos a proceder a realizar esta misma petición pero usando la librería **request**.

```python
#Importamos la librería request
import requests

#Especificamos la url donde queremos realizar la petición 
url = "http://www.datacamp.com/teach/documentation"

#Hacemos la petición y capturamos la respuesta
response = requests.get(url)

#Extraemos la respuesta
text = response.text

#Vemos el resultado
print(text)
```

# Parsing HTML with BeautifulSoup

A continuación aprenderemos a como usar el paquete **BeautifulSoup** para parsear, embellecer y extraer información de ficheros HTML.

```python
#Importamos librerías
import requests 
from bs4 import BeautifulSoup

#Especificamos la url que queremos parsear
url = 'https://www.python.org/~guido/'

#Hacemos la petición a la url y capturamos la respuesta
response = requests.get(url)

#Extraemos la respuesta
html_doc = response.text

#Creamos un objeto BeautifulSoup
soup = BeautifulSoup(html_doc)

#Ponemos en un formato más adecuado el archivo html
pretty_soup = soup.prettify()

#Vemos el resultado 
print(pretty_soup)
```

# Turning a webpage into data using BeautifulSoup: getting the text

A continuación, aprenderemos como extraer información relevante de nuestro archivo HTML.

```python
#Extraemos el título de página web
page_title = soup.title

#Mostramos el resultado
print(page_title)
```

```python
#Extraemos el texto
page_text = soup.text

#Vemos el resultado
print(page_text)
```

# Turning a webpage into data using BeautifulSoup: getting the hyperlinks

A continuación veremos como extraer hyperlinks de un archivo HTML. Para ello vamos hacer uso del método **find_all()**

```python
#Encontramos todos los tags con 'a' que es la etiqueta que define un hyperlink en html
a_tags = soup.find_all('a')

#Ahora procedemos a mostrar los hyperlinks
for link in a_tags:
    print(link.get('href'))
```
