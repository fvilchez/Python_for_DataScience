```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

# Generating meshes

En lugar de visualizar datos de arrays de dos dimensiones, es necesario comprender como se generan y se manipulan arrays 2D. A continuación vamos a hacer uso de la función **meshgrid** de la librería Numpy para generar arrays de dos dimensiones que pueden ser visualizadas mediante **plt.imshow()**

```python
#Generamos un meshgrid
Y, X = np.meshgrid(range(10), range(20))
```

El código anterior genera dos arrays con 20 filas y 10 columnas, que corresponden a 20 filas a lo largo del eje Y y 10 columnas a lo largo del eje X.

```python
#Nos creamos  dos arrays unidemensiones 
u = np.linspace(-2, 2, 41)
v = np.linspace(-1, 1, 21)

#Nos generamos dos arrays bidimensionales
X, Y = np.meshgrid(u,v)

#Obtenemos el valor que queremos representar 
Z = np.sin(3*np.sqrt(X**2 + Y**2))

#Mostramos el resultado
plt.pcolor(Z)
plt.show()
```

# Contour & filled contour plots

A pesar de que **plt.imshow()** y **plt.pcolor()** son funciones a menudo usadas para visualizar arrays 2D, tenemos otras formas de visualizar este tipo de datos. Una forma es la de crear contornos a partir de los valores de esta array. Dos tipos de contornos son soportados por Matplotlib: **plt.contour()** y **plt.contourf()**, la primera de ellas muestra los contornos como líneas y la otra muestra los contornos rellenos.

```python
#Nos generamos el contorno básico
plt.subplot(2,2,1)
plt.contour(X, Y, Z)

#Nos generamos el contorno de líneas con 20 contornos
plt.subplot(2,2,2)
plt.contour(X, Y, Z, 20)

#Nos generamos un plot de contornos rellenos
plt.subplot(2,2,3)
plt.contourf(X, Y, Z)

#Nos generamos el plot de contornos rellenos con 20 contornos
plt.subplot(2,2,4)
plt.contourf(X, Y, Z, 20)

#Separamos los plots
plt.tight_layout()
plt.show()
```

# Modifying colormaps

Cuando mostramos un array 2D con **plt.imshow()** o **plt.pcolor()**, los valores de nuestra array son mapeados al color que le corresponda. El conjunto de colores usado es determinado por el colormap que asigna de forma suavizada un color a los valores, lo que facilita la compresión de como están los datos estructurados de un solo vistazo.
En determinadas situaciones es posible querer cambiar el colormap asignado por defecto (**jet**). Un buen mapa de colores debe ser agradable visualmente y además representar de forma clara como están estructurados nuestros datos.

* Algunos colormaps tienen nombres únicos como jet, coolwarm, magma, viridis.

* Otros se tratan de esquemas basados en colores Greens, Blues, Reds o Purples.

* Otros tienen nombres basados en estaciones del año, summer, autumn, winter, spring.

Para cambiar el tipo de colormap a utilizar disponemos del parámetro **cmap**.

```python
#Nos creamos un mapa de contorno relleno con un mapa de colores viridis
plt.subplot(2, 2, 1)
plt.contourf(X, Y, Z, 20, cmap = 'viridis')
plt.colorbar()
plt.title('viridis')

#Nos creamos un mapa de contorno rellenos con un mapa de colores gray
plt.subplot(2,2,2)
plt.contourf(X, Y, Z, 20, cmap = 'gray')
plt.colorbar()
plt.title('gray')

#Nos creamos un mapa de contorno relleno con un mapa de colores autumn
plt.subplot(2,2,3)
plt.contourf(X, Y, Z, 20,cmap = 'autumn')
plt.colorbar()
plt.title('autumn')

#Nos creamos un mapa de contorno relleno con un mapa de colores winter
plt.subplot(2,2,4)
plt.contourf(X, Y, Z, 20,cmap = 'winter')
plt.colorbar()
plt.title('winter')

#Separamos los plots
plt.tight_layout()
plt.show()
```

# Using hist2d()

Dado un conjunto de pares ordenados, podemos contar el número de puntos con valores similares haciendo uso de un histograma de dos dimensiones. Es similar a un histograma de una sola dimensión, pero en este caso describe la variación de dos variables en lugar de una sola variable. En matplotlib contamos con la función **plt.hist2d()**:

* Como primer parámetro le pasamos las arrays de las cuales queremos ver su distribución, estas dos arrays deben de tener la misma longitud.

* Podemos especificar el número de bins bins = (nx, ny) donde nx es el número de bins para la variable x y ny hace referencia al número de bins de la variable y.

* Podemos limitar la región rectangular donde se van a representar nuestros datos, para ello contamos con            range = ((xmax, xmin), (ymax,ymin))

```python
#Cargamos el conjunto de datos
auto_mpg = pd.read_csv('auto-mpg.csv')

#Extraemos los datos que queremos representar
mpg = auto_mpg.loc[:, 'mpg']
hp = auto_mpg.loc[:, 'hp']

#Nos creamos el histogramada 2d
plt.hist2d(hp, mpg, bins = (20, 20), range = ((40, 235), (8,48)))

#Añadimos la paleta de colores
plt.colorbar()

#Añadimos descripción de los ejes
plt.xlabel('Horse power (hp)')
plt.ylabel('Miles per gallon (mpg)')

#Mostramos el resultado
plt.show()
```

# Using hexbin()

La función **plt.hist2d()** hace uso de cajas rectangulares a la hora de construir nuestro histograma de dos dimensiones. La función **plt.hexbin()** usa cajas hexagonales.

* El parámetro gridsize que por defecto toma el valor 100 nos permite determinar el númeor de hexágonos tanto para el eje x como para el eje y, se le pasa en forma de tupla.

* El parámetro extent = (xmin, xmax, ymin, ymax) nos permite definir el rango en el cual se moverano nuestras variables.

```python
#Nos generamos nuestro grafo 
plt.hexbin(hp, mpg, gridsize=(15,12), extent=(40, 235, 8, 48))

#Agregamos los ejes
plt.xlabel('Horse power (hp)')
plt.ylabel('Miles per gallon (mpg)')
plt.colorbar()

#Mostramos el resultado 
plt.show()
```

# Loading, examining images

Las imágenes a color tales como fotografías contienen la intensidad de los colores rojo, verde y azul.

```python
#Cargamos la imagen
img = plt.imread('600px-Astronaut-EVA.jpg')

#Vemos el resultado 
plt.imshow(img)
plt.show()
```

Podemos eliminar los valores de los ejes haciendo uso de **plt.axis('off')**

```python
#Cargamos la imagen
img = plt.imread('600px-Astronaut-EVA.jpg')

#Vemos el resultado 
plt.imshow(img)
plt.axis('off')
plt.show()
```

# Pseudocolor plot from image data

Existen situaciones en las que no deseamos mostrar los canales RGB.

```python
#Calculamos las intensidades de cada uno de los canales de nuestra matriz
intensity = img.sum(axis = 2)

#Visualizamos el resultado
plt.imshow(intensity, 'gray')
plt.colorbar()
plt.axis('off')
plt.show()
```
