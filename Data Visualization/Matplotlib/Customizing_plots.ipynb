```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Multiple plots on single axis

Matplotlib nos permite representar múltiples plots en un mismo eje. Para esto bastará con hacer uso de **plt.plot()**.

```python
#Cargamos los datos
women_bachelor_USA = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_bachelor_USA.head()
#Nos creamos un primer plot con el número de mujeres por año graduadas en física
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Physical Sciences'], color = 'blue')

#En el mismo eje nos creamos un segundo plot con el número de mujeres graduadas en computer science
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Computer Science'], color = 'red')

#Mostramos el resultado
plt.show()
```

# Using Axes

En lugar de superponer plot en ejes comunes, podemos preferir plotear estos plots en distintos ejes. El comando **plt.axes()** es una forma de poder hacer esto (pero requiere especificar coordenadas relativas al tamaño de la figura). En la llamada a **plt.axes()** les  pasamos las coordenadas de la esquina inferior izquierda, la anchura y la altura de nuestra figura, estas coordenadas deben pasarse en forma de lista o tupla. Son valores entre 0 y 1 representando longitudes relativas a la dimensión de la figura.

```python
#Nos creamos el primer axes
plt.axes([0.05, 0.05, 0.425, 0.9])

#Mostramos el número de mujeres graduadas en físicas
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Physical Sciences'], color = 'blue')

#Nos creamos un segundo eje
plt.axes([0.525, 0.05, 0.425, 0.9])

#Mostramos el número de mujeres graduadas en Computer Science
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Computer Science'], color = 'red')

#Mostramos el resultado
plt.show()
```

# Using subplot() (1)

El comando **plt.axes()** requiere de un esfuerzo a la hora de usarse, ya que las coordenadas deben de ser puestas de forma manual. Una mejor alternativa es de la usar **plt.subplot()**, ya que nos permite determinar el diseño de forma automática. Al comando **plt.subplot()** le pasamos tres argumentos principales. Los dos primeros son la dimensión de nuestro **grid**, es decir, si le pasamos los valores 2,2 esto indicará que hacemos un grid de 2x2, es decir, podemos representar hasta 4 plots. El tercer parámetro es el plot activo (de izquierda a derecha y de arriba a bajo).

```python
#Nos creamos un subplot de tamaño 1x2 y activamos el primer plot
plt.subplot(1,2,1)

#Nos creamos un plot que indique el número de mujeres que se licenciarion en físicas
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Physical Sciences'], color = 'blue')

#Hacemos uso del comando plt.title para poner título a nuestro plot
plt.title('Physical Sciences')

#Activamos el segundo plot de nuestro grid
plt.subplot(1,2,2)

#Nos creamos un plot que indique el número de mujeres que se licenciarion en Computer Science
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Computer Science'], color = 'red')

#Hacemos uso del comando plt.title para poner título a nuestro plot
plt.title('Computer Science')

#Hacemos uso del comando plt.tight_layout para poner una sepación adecuada entre plots
plt.tight_layout()

#Mostramos el resultado
plt.show()
```

# Using subplot() (2)

A continuación vamos a proceder a hacer uso de subplot para generar un grid de tamaño 2x2.

```python
#Nos generamos el grid y 'activamos el primer plot
plt.subplot(2,2,1)

#Representamos el número de mujeres que se graduaron en Physical Sciences
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Physical Sciences'], color = 'blue')

#Ponemos el título a nuestro plot
plt.title('Physical Sciences')

#Activamos el segundo plot
plt.subplot(2,2,2)

#Representamos el número de mujeres que se graduaron en Computer Science
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Computer Science'], color = 'red')

#Ponemos el título a nuestro plot
plt.title('Computer Science')

#Activamos el tercer plot
plt.subplot(2,2,3)

#Vemos el número de mujeres que se graduaron en Engineering
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Engineering'], color = 'green')

#Ponemos el título a nuestro tercer plot
plt.title('Engineering')

#Activamos el cuarto plot
plt.subplot(2,2,4)

#Vemos el número de mujeres que se graduaron en Mathematics and Statistics
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Math and Statistics'], color = 'yellow')

#Ponemos el título a nuestro plot
plt.title('Math and Statistics')

#Hacemos del tight_layout
plt.tight_layout()

#Mostramos los resultados
plt.show()
```

# Using xlim(), ylim()

Los comandos **plt.xlim()** y **plt.ylim()** nos permiten hacer zoom en nuestro gráfico en aquellas zonas que estemos interesados. 

```python
#Generamos un plot con el número de mujeres graduadas en Physical Sciences y Computer Science en un mismo gráfico
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Physical Sciences'], color = 'red')
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Computer Science'], color = 'blue')

#Agregamos los nombres de los ejes
plt.xlabel('Year')
plt.ylabel('% of women')

#Hacemos zoom entre los años 1990 y 2010
plt.xlim((1990, 2010))

#Hacemos zoom en el eje y entre 0 y 50
plt.ylim((0,50))

#Le ponemos un título a nuestro plot
plt.title('Degrees awarded to women (1990-2010)\nComputer Science(blue)\nPhysical Sciences(red)')

plt.show()
```

# Using axis()

El comando **plt.axis()** nos permite fijar estos límites de una sola vez. En primer lugar le pasamos los límites del eje x y seguidamente los del eje y.

```python
#Generamos un plot con el número de mujeres graduadas en Physical Sciences y Computer Science en un mismo gráfico
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Physical Sciences'], color = 'red')
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Computer Science'], color = 'blue')

#Agregamos los nombres de los ejes
plt.xlabel('Year')
plt.ylabel('% of women')

#Hacemos zoom con plt.axis()
plt.axis((1990, 2010, 0, 50))

#Le ponemos un título a nuestro plot
plt.title('Degrees awarded to women (1990-2010)\nComputer Science(blue)\nPhysical Sciences(red)')

plt.show()
```

# Using legend()

Cuando tenemos múltiples plots en un mismo axes, la leyenda nos permite distinguir de forma clara cada uno de estos plots. A la hora de integrar esta leyenda en matplotlib tenemos que seguir dos pasos:

* Etiquetar la figura: para esto disponemos del argumento **label** dentro de plt.plot().

* Hacer uso de comando lengend(), indicando mediante el argumento loc donde queremos colocar nuestra leyenda.

El argumento **loc** puede tomar los siguientes valores:

* upper left

* center left

* lower left

* best

* upper center

* center

* lower center

* upper right

* lower right

* center right

* right

```python
#Generamos un plot con el número de mujeres graduadas en Physical Sciences y Computer Science en un mismo gráfico
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Physical Sciences'], color = 'red', label = 'Physical Sciences')
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Computer Science'], color = 'blue', label = 'Computer Science')

#Agregamos los nombres de los ejes
plt.xlabel('Year')
plt.ylabel('% of women')

#Hacemos zoom con plt.axis()
plt.axis((1990, 2010, 0, 50))

#Fijamos la leyenda 
plt.legend(loc = 'best')

#Le ponemos un título a nuestro plot
plt.title('Degrees awarded to women (1990-2010)')

plt.show()
```

# Using annotate()

En determinadas ocasiones puede ser útil realizar algún tipo de anotación sobre un plot. Esto puede hacer que nuestros plots sean más interpretables. Matplotlib nos permite realizar este tipo de anotaciones mediante texto y mediante gráficos de tipo flecha.

```python
#Nos creamos el plot
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Physical Sciences'], color = 'red', label = 'Physical Sciences')
plt.plot(women_bachelor_USA.loc[:, 'Year'], women_bachelor_USA.loc[:, 'Computer Science'], color = 'blue', label = 'Computer Science')

#Nombramos los ejes y el título
plt.xlabel('Year')
plt.ylabel('% of women')
plt.title('Degrees awarded to women (1990-2010)')

#Fijamos la leyenda
plt.legend(loc = 'lower center')

#Calculamos el máximo de graduaciones en computer science y en que año se produjo dicho máximo
cs_max = women_bachelor_USA.loc[:, 'Computer Science'].max()
yr_max = women_bachelor_USA.loc[:, 'Year'][ women_bachelor_USA.loc[:, 'Computer Science'].argmax()]

#Agregamos las anotaciones
plt.annotate('Maximum CS', xy = (yr_max, cs_max), xytext = (yr_max + 5, cs_max + 5),
            arrowprops = dict(facecolor = 'black'))

plt.show()
```
