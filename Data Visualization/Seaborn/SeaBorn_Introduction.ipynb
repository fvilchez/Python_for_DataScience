```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Comparing a histogram and distplot

La librería **seaborn** nos permite realizar un análisis más estadístico y nos puedo proporcionar insights más sofisticados de nuestros datos.

```python
#Cargamos datos
df = pd.read_csv('schoolimprovement2010grants.csv')

#Nos creamos el grid y mostramos el histograma de la columna Award_Amount
plt.subplot(1,2,1)
df.Award_Amount.hist()

#Nos creamos el distplot con seaborn
plt.subplot(1,2,2)
sns.distplot(df.Award_Amount)

plt.tight_layout()
plt.show()
```

Podemos ver que ambos gráficos son complementarios, es decir, ambos representan la distribución de nuestros datos.

# Plot a histogram

La función **distplot()** retornar por defecto la función Kernel Density Gaussian (KDE). La función KDE nos permite suavizar nuestra distribución y es una forma bastante últil a la hora de ver nuestros datos. Sin embargo **seaborn** también nos permite realizar un histograma básico si es esto lo que necesitamos. Para ello basta con poner a **False** el argumento **kde** de nuestro función **distplot()**.

```python
#Hacemos un histograma básico mediante seaborn
sns.distplot(df.Award_Amount, 
            kde = False,
            bins = 20)

plt.show()
```

# Rug plot and kde shading

Podemos agregar un gráfico de tipo rug a nuestro plot. Este tipo de gráficos nos pueden ayudar a ver información que quizás se nos esté escapando en otro tipo de gráficos. Se trata de un gráfico que representa puntos mediante líneas.

```python
#Nos creamos el gráfico
sns.distplot(df.Award_Amount,
            hist = False,
            rug = True,
            kde_kws = {'shade': True})

plt.show()
```

# Create a regression plot

Los plots de regresión nos permite realizar un análisis más claro de nuestro conjunto de datos. Este tipo de gráficos nos permiten ver si realmente existe una relación lineal entre dos variables. A la hora de hacer este tipo de gráficos disponemos de **regplot()** y de **lmplot()**.

```python
#Cargamos los datos
df = pd.read_csv('insurance_premiums.csv')

#Nos creamos nuestro plot en forma de regresión
sns.regplot(data = df,
           x = 'insurance_losses',
           y = 'premiums')

plt.show()
```

```python
#Hacemos lo mismo haciendo uso de lmplot
sns.lmplot(data = df,
          x = 'insurance_losses',
          y = 'premiums')

plt.show()
```

Aunque los resultados aparentemento son iguales, veremos como **lmplot()** se trata de una función más potente, ya que disponde de una serie de argumentos que la hacen más potente.

# Plotting multiple variables

La función **lmplot()** nos permite ver insights más sofisticadas de nuestros datos. Por ejemplo supongamos un conjunto de datos que contiene información por estado y área. Podría ser interesante ver si existe una difencia entre regiones del país.

```python
#Nos creamos el grafo
sns.lmplot(data = df, 
          x = 'insurance_losses',
          y = 'premiums',
          hue = 'Region')

plt.show()
```

En caso de que este tipo de gráfico no nos quede muy claro, podemos hacer uso del argumento **col** que nos permite mostrar esto mismo en plots por separado.

```python
#Nos creamos el grafo 
sns.lmplot(data = df,
          x = 'insurance_losses',
          y = 'premiums',
          col = 'Region')

plt.show()
```
