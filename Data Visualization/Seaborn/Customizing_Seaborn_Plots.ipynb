```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Setting the default style

A continuación vamos a realizar un análisis de los precios de las casas calculados por el departamente de vivienda y desarrollo de los Estados Unidos. Los precios varían mucho a lo largo de USA.

```python
#Cargamos los datos
df = pd.read_csv('FY18_4050_FMRs.csv')
df.head(3)
```

```python
#Visualizamos el histograma por defecto
df.fmr_2.plot.hist()
plt.title('Normal style')
plt.show()
```

```python
#Visualizamos el histograma poniendo style seaborn
sns.set()
df.fmr_2.plot.hist()
plt.title('Seaborn style')
plt.show()
```

# Comparing styles

Seaborn soporta una gran cantidad de estilos, podemos controlar este tipo de estilos mediante el comando **sns.set_style()**.

```python
#Nos creamos un distplot con el estilo dark
sns.set_style('dark')
sns.distplot(df.fmr_2)
plt.title('dark style')
plt.show()
```

```python
#Nos creamos un distplot con el estilo whitegrid
sns.set_style('whitegrid')
sns.distplot(df.fmr_2)
plt.title('whitegrid style')
plt.show()
```

# Removing spines

En general, las visualizaciones deberían de minimizar marcas extrañas, es decir, los datos deberían de hablar por sí solos. Seabonr nos permite eliminar las líneas tanto en la parte superior, inferior, izquierda o derecha, que usualmente se suelen llamar spines.

```python
#Nos creamos un plot de tipo lmplot con el fin de comparar las variables pop2010 y fmr_2

#Fijamos el estilo
sns.set_style('white')
#Nos creamos el lmplot
sns.lmplot(data = df,
          x = 'pop2010',
          y = 'fmr_2')
#Eliminamos spines top y right
sns.despine(top = True, right = True)

#Mostramos el resultado
plt.show()
```

# Matplotlib color codes

Seaborn ofrece una gran cantidad de opciones para modificar los colores. La forma más simple es especificar el color en el plot.  Para ello disponemos del comando **sns.set(color_codes = True)**. Tras esto podemos hacer uso del parámetro color en nuestros gráficos y seleccionar el color deseado. Una forma fácil es usar el mismo sistema de colores que usa matplotlib.

```python
#Activamos colores
sns.set(color_codes = True)
#Nos creamos un distplot de color magenta
sns.distplot(df.fmr_3, color = 'm')
#Mostramos el resultado
plt.show()
```

# Using default palettes

Seaborn también incluye una gran variedad de paletas de colores que pueden sar facilmente aplicadas a nuestros plots. Para fijar la paleta de colores que queremos usar disponemos del comando **sns.set_palette()**.

```python
for p in ['bright', 'colorblind']:
    sns.set_palette(p, 2)
    sns.distplot(df.fmr_3)
    plt.show()
```

La función **sns.palplot()** nos permite mostrar una paleta, mientras que la función **sns.color_palette()** retorna la paleta actual.

```python
for p in sns.palettes.SEABORN_PALETTES:
    print(p)
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.show()
```

# Creating Custom Palettes

Seleccionar una paleta adecuada para tu conjunto de datos puede conllevar un tiempo bastante grande. Afortunadamente, seaborn dispone de una serie de funciones que nos permite crear nuestra propia paleta de colores de forma secuencial, categorica o divergete. Seaborn nos permite visualizar distintas paletas de forma muy sencilla con la función **palplot()**.

```python
#Nos creamos una paleta de colores violetas
sns.palplot(sns.color_palette('Purples', 8))
plt.show()
```

```python
#Nos creamos una paleta de 10 colores haciendo uso de husl
sns.palplot(sns.color_palette('husl', 10))
plt.show()
```

```python
#Nos creamos una paleta de 6 colores de tipo coolwarm
sns.palplot(sns.color_palette('coolwarm', 6))
plt.show()
```

# Using matplotlib axes

Seaborn hace uso de matplotlib como librería base a la hora de crear plots. En la gran mayoría de los casos podemos hacer uso de seaborn para modificar nuestros gráficos pero en otras situaciones puede ser útil hacer uso de matplotlib para personalizar nuestros gráficos. En este caso en elemento fundamental son los elementos de tipo **axes**. Una vez tenemos creado un objeto de tipo **axes**, podemos realizar personalizar nuestro plot de una gran forma.

```python
#Nos creamos nuestra figura y nuestro axes
fig, ax = plt.subplots()

#Ahora nos creamos nuestro plot
sns.distplot(df.fmr_3, ax = ax)

#Agragamos nombre al eje x
ax.set(xlabel = '3 bedroom Fair Market Rent')

#Mostramos el resultado
plt.show()
```

# Additional plot customizations

```python
#Nos creamos nuestra figura y nuestro axes
fig, ax = plt.subplots()

#Ahora nos creamos nuestro gráfico
sns.distplot(df.fmr_1, ax = ax)

#Modificamos el aspecto
ax.set(xlabel = '1 bedroom Fair Market rent',
      xlim = (100, 1500),
      title = 'US Rent')

#Mostramos el resultado 
plt.show()
```

# Adding annotations

```python
#Nos creamos nuestra figura y axes
fig, ax = plt.subplots()

#Nos creamos el gráfico
sns.distplot(df.fmr_1, ax = ax)

#Agregamos nombre a los ejes, agregamos título, límite en el eje x
ax.set(xlabel = '1 Bedroom Fair Market Rent', xlim = (100, 1500), title = 'US Rent')

#Agregamos una línea vertical en la mediana y otra en la media
ax.axvline(x = df.fmr_1.median(), color = 'm', label = 'median', linestyle = '--', linewidth = 2)
ax.axvline(x = df.fmr_1.mean(), color = 'b', label = 'mean', linestyle = '-', linewidth = 2)

#Agregamos la leyenda
ax.legend()

#Vemos el resultado 
plt.show()
```

# Multiple plots

```python
#Nos creamos nuestra objeto tipo figure y nuestros objetos tipo axes
fig, (ax0, ax1) = plt.subplots(nrows = 1, ncols = 2, sharey = True)

#Nos creamos nuestro distplot para casas con una habitación
sns.distplot(df.fmr_1, ax = ax0)
ax0.set(xlabel = '1 Bedroom Fair Market Rent', xlim  = (100,1500))

#Nos creamos nuestro distplot para casas con dos habitaciones
sns.distplot(df.fmr_2, ax = ax1)
ax1.set(xlabel = '2 Bedromm Fair Market Rent', xlim = (100, 1500))

#Mostramos el resultado
plt.show()
```
