```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Carga de datos
df = pd.read_csv('schoolimprovement2010grants.csv')
```

# Categorical Plots

## `stripplot()` and `swarmplot()`

Cuando trabajamos con conjunto de datos, es muy probable que nos encontremos que estos conjuntos de datos además de variables numéricas contengan variables categóricas. Para la representación gráfica de estos conjuntos de datos, o para poder representar valores numéricos en función de valores categóricos la librería **seaborn** dispone de una gran variedad de gráficos.

```python
#Nos generamos nuestro gráfico stripplot
plt.figure(figsize=(10,6))
sns.stripplot(data = df, 
             x = 'Award_Amount',
             y = 'Model Selected',
             jitter = True)

plt.show()
```

Este gráfico nos permite ver la variable numérica **Award_Amount** para los distintos valores de la variable categórica **Model Selected**. Este gráfico no llega a ser del todo claro, para este tipo de gráficos disponemos del gráfico **swarmplot**, este gráfico hace uso por debajo de un algoritmo más complejo que nos permite visualizar la información mejor. El principal problema de los gráficos **swarmplot** es que no escalan muy bien datasets pesados.

```python
#Nos generamos nuestro gráfico swarmplot
plt.figure(figsize=(10,6))
sns.swarmplot(data = df,
             x = 'Award_Amount',
             y = 'Model Selected',
             hue = 'Region')
plt.show()
```

En este grafo podemos ver como los puntos son representados de forma más clara, además el parámetro **hue** nos permite colorear los puntos en función de otra variable categórica.

## boxplots, violinplots and lvplots

Los **boxplots** o gráficos de caja nos permiten ver la distribución que siguen nuestros datos, mediante métricas como la mediana o los quartiles. Además nos permite ver si tenemos outliers en nuestros datos.

```python
#Nos creamos nuestro gráfico de tipo boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data = df, 
           x = 'Award_Amount',
           y = 'Model Selected')
plt.show()
```

Los **violinplots** se tratan de un tipo de plot alternativo al boxplot, este gráfico hace uso de la KDE no se muestran todos los puntos de nuestros datos. Cuando estamos trabajando con datos multimodales, los gráficos de tipo boxplot se ven muy limitados, y es en este caso donde los gráficos de tipo **violinplots** son realmente útiles.

```python
#Nos creamos nuestro gráfico de tipo violinplot
plt.figure(figsize=(10,6))
sns.violinplot(data = df, 
              x = 'Award_Amount',
              y = 'Model Selected',
              palette = 'husl')
plt.show()
```

Los elementos de este gráfico **violinplot** se interpretan de la siguiente manera:

* El punto blanco representa la mediana

* La barra gris gruesa en el centro representa el rango intercuatílico

* La barra gris delgada representa el intervalo de confianza del 95%

Las zonas anchas representan una elevada probabilidad de que nuestra muestra tome esos valores, mientras que las zonas más estrechas representan una probabilidad baja de que nuestra población tome esos valores. Se trata de un gráfico muy útil para representar conjunto de datos pesados, aunque puede tardar en crearse.

Los gráficos **lvplot** se tratan de gráficos que son una fusión entre los **violinplot** y los **boxplot** pero que son más recomendables a la hora de trabajar con datasets pesados y además son muy intuitivos y fáciles de interpretar.

```python
#Nos creamos nuestro gráfico de tipo lvplot
plt.figure(figsize=(10,6))
sns.lvplot(data = df, 
           x = 'Award_Amount',
           y = 'Model Selected',
           palette = 'Paired',
           hue = 'Region')
plt.show()
```

## barplots, pointplots and countplots

Los gráficos de **countplot** se tratan de gráficos que nos permite mostrar el número de veces que apareces cada una de los valores de una variable categórica.

```python
#Nos creamos un gráfico de countplot
plt.figure(figsize=(10,6))
sns.countplot(data = df,
             y = 'Model Selected',
             hue = 'Region')

plt.show()
```

Los gráficos de tipo **pointplot** se tratan de gráficos que nos permiten mostrar una estimación de la tendencia central para una variable numérica. Este tipo de gráficos pueden ser más efectivos que los gráficos de tipo barra si queremos comparar diferentes niveles de una o más variables categóricas. Son realmente buenos gráficos a la hora de mostrar interacciones: como los distintos niveles de una variale categórica cambian a través de los segundos niveles de otra variable categórica. Las líneas que unen cada punto permiten que las interacciones sean juzgadas a partir de una pendiente que puede llegar a ser más clara que alturas de barras. Los bigotes muestran una estimación del posible error cometido a la hora de estimar la tendencia central.

```python
#Nos creamos nuestros gráficos de tipo pointplot
plt.figure(figsize=(10,6))
sns.pointplot(data = df,
             y = 'Award_Amount',
             x = 'Model Selected',
             capsize = .1)
plt.show()
```

Finalmente los gráficos de de tipo **barplot** también nos permiten realizar una estimación central de nuestra variable con la altura de las barras proporcionando además presenta una estimación del error cometido en la estimación de la tendencia central.

```python
#Nos creamos nuestro gráfico de tipo barplot
plt.figure(figsize=(10,6))
sns.barplot(data = df,
           y = 'Award_Amount',
           x = 'Model Selected',
           hue = 'Region')

plt.show()
```

# Regression and residual plots

La regresión lineal se trata de una herramienta realmente útil a la hora de comprender la relación entre variables numéricas. Seaborn dispone de simples pero poderosas herramientas para examinar estas relaciones. Una de las funciones es **regplot**, esta nos pinta el modelo de regresión adaptado a los datos.

```python
#Cargamos los datos
df = pd.read_csv('college_datav3.csv')

#Nos creamos nuestro plot de regresión
plt.figure(figsize=(10,6))
sns.regplot(data = df,
           x = 'SAT_AVG_ALL',
           y = 'Tuition',
           marker = '^',
           color = 'g')
plt.show()
```

Otra de las funciones es **residplot** esta función  nos pinta los residuos tras aplicar la regresión, si la relación entre variables es lineal los residuos deberían estar dispersados de forma aleatoria respecto al eje horizontal.

```python
#Nos creamos nuestra gráfico residplot
plt.figure(figsize=(10,6))
sns.residplot(data = df,
             x = 'SAT_AVG_ALL',
             y = 'Tuition')
plt.show()
```

## Regression plot parameters

Los gráficos de tipo regresivos permite hacer uso de múltiples parámetros que pueden ser usados para configurar estos gráficos y encontrar más información respecto a los datos. Por ejemplo podemos tratar de lanzar una regresión por una variable categórica.

```python
#Nos creamos una regresión a partir de una categórica
plt.figure(figsize=(10,6))
sns.regplot(data = df,
           y = 'Tuition',
           x = 'REGION')

plt.show()
```

Puesto que estamos ante una variable discreta, podemos hacer uso del parámetro **x_estimator** este parámetro aplica la función indicada a cada uno de los conjuntos que se crean a partir de nuestra variable discreta. 

```python
plt.figure(figsize=(10,6))
sns.regplot(data = df,
           y = 'Tuition',
           x = 'REGION',
           x_estimator = np.mean)

plt.show()
```

También podemos indicar el orden del polinomio que queremos aplicar, para ello contamos con el parámetro **order**.

```python
plt.figure(figsize=(10,6))
sns.regplot(data = df,
           y = 'Tuition',
           x = 'REGION',
           x_estimator = np.mean,
           order = 3)

plt.show()
```

## Binning data

CUando los datos del eje x son valores continuos, puede ser útil separar estos en varios bins con el objetivo de tener una mejor visualización de los cambios en los datos. Uno de los parámetros de los que dispone nuestro gráfico **regplot** se trata del parámetro **fit_reg** puede tomar el valor de True o False. Por defecto toma el valor de True en caso de que le asignemos el valor de False no será fijada la linea de regresión y simplemente se procederán a representar los datos.

```python
#Nos creamos nuestro regplot con fit_reg igual a False
plt.figure(figsize=(10,6))
sns.regplot(data = df,
           y = 'Tuition',
           x = 'UG',
           fit_reg = False)

plt.show()
```

Ahora procedemos a hacer uso del parámetro **bins**. Esta función crea tantos bins como le indiquemos y lo que hace es calcular para cada uno de estos bins la tendencia central y un intervalo de confianza. Este agrupamiento solo influye en la forma en la cual se representa el diagrama de dispersión, la recta de regresión se sigue ejecutando sobre los datos originales.

```python
#Nos creamos nuestro regplot con x_bins igual a 5
plt.figure(figsize=(10,6))
sns.regplot(data = df,
           y = 'Tuition',
           x = 'UG',
           x_bins = 5)

plt.show()
```

```python
#Nos creamos nuestro regplot con x_bins igual a 8
plt.figure(figsize=(10,6))
sns.regplot(data = df,
           y = 'Tuition',
           x = 'UG',
           x_bins = 8)

plt.show()
```

# Matrix plots

## Creating heatmaps

Los mapas de calor se tratan de gráficos que pueden ser muy útiles a la hora de sumarizar la relación entre dos variables. Este tipo de gráficos requieren que le pasemos los datos de una forma adecuada, para esto disponemos de la función **crosstab** de la librería panas.

```python
#Cargamos los datos
df = pd.read_csv('daily_show_guests_cleaned.csv')

#Convertimos los datos a un formato adecuado
pd_crosstab = pd.crosstab(df["Group"], df["YEAR"])

#Nos creamos el heatmap
plt.figure(figsize=(10,6))
sns.heatmap(pd_crosstab)

plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
```

## Customizing heatmaps

Los gráficos de tipo **heatmap** disponen de una gran variedad de argumentos que nos permiten personalizar nuestro gráfico.

```python
# Create the crosstab DataFrame
pd_crosstab = pd.crosstab(df["Group"], df["YEAR"])

# Plot a heatmap of the table with no color bar and using the BuGn palette
plt.figure(figsize=(10,6))
sns.heatmap(pd_crosstab, cbar=False, cmap="BuGn", linewidths=.3, annot = True)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

#Show the plot
plt.show()
```
