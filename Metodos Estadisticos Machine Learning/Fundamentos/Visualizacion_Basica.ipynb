```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import seed, randint, randn
```

# Introducción 

En la gran mayoría de las ocasiones los datos no tienen sentido hasta que no son visualizados. Ser capaz de visualizar datos de una forma adecuada se trata de un skill que todo Data Science debe tener.

# Gráficos de Línea

Un gráfico de línea es generalmente usado para representar observaciones recogidas en intervalos regulares. El eje x representa el intervalo regular, por ejemplo, el tiempo. El eje y muestra las observaciones, ordenadas por el eje x y conectadas por una línea.

Los gráficos de línea son útiles para representar series temporales así como cualquier secuencia donde existe un orden en las observaciones.

Python disponde de la librería **matplotlib** que es una entre muchas de las librerías que nos permite realizar visualizaciones. Se trata de la librería principal de Python y esta librería es usada de forma subyacente por otras librerías como **seaborn**.

A la hora de realizar gráficos de línea podemos hacer uso del comando **plot()**, pasándole en su forma más básica como primer parámetro los valores del eje x y como segundo parámetro los valores del eje y.

```python
#Generamos un intervalo para el eje x
x = [x*0.1 for x in range(100)]

#Generamos un intervalo para el eje y 
y = np.sin(x)

#Creamos y visualizamos nuestro gráfico de línea
plt.plot(x, y)
plt.show()
```

# Gráficos de Barras

Un gráfico de barras es usado generalmente para presentar cantidades relativas para múltiples categorías. El eje x representa las categorías y están espaciadas por igual. El eje y representa la cantidad para cada categoría y es dibujada como una barra desde la base hasta el nivel apropiado en el eje y. Los gráficos de barra pueden ser útiles para comparar múltiples puntos cuantitativos o estimaciones.

Mediante la librería matplotlib, podemos realizar gráficos de barras mediante la función **bar()**. Esta función en su forma básica recibe como parámetros los valores del eje x y los del eje y.

```python
#Fijamos la semilla 
seed(1)

#Fijamos el eje x
x = ['red', 'green', 'blue']

#Fijamos el eje y 
y = [randint(0,100), randint(0,100), randint(0,100)]

#Creamos y visualizamos nuestro gráfico de barras
plt.bar(x, y)
plt.show()
```

# Histogramas

Los histogramas son usados generalmente para resumir la distribución de una muestra de datos. El eje x representa cajas o intervalos discretos para las observaciones. Por ejemplo, observaciones con valores entre 1 y 10 pueden ser espaciados en 5 intervalos, los valores entre [1,2] perteneceran al primer intervalo, los valores entre [3,4] perteneceran al segundo intervalo etc. El eje y representa la frecuencia o el conteo del número de observaciones en el conjunto de datos que pertenecen a cada caja.

Podemos hacer uso de la función **hist()** perteneciente a la librería matplotlib a la hora de visualizar un histograma. En este caso en su forma básica recibe como parámetro la variable de la cual queremos visualizar su histograma y el número de bins o intervalos que queremos visualizar.

```python
#Fijamos la semilla
seed(1)

#Generamos una variable 
x = randn(1000)

#Generamos y visualizamos el histograma
plt.hist(x, bins = 20, ec = 'black')
plt.show()
```

# Diagramas de Caja o Box Plot

Un diagrama de cajas, más conocido como BoxPlot, se trata de un gráfico que es usado genralmente para resumir la distribución de una muestra de datos. El eje X representa la muestra de datos, donde múltiples diagramas de cajas pueden ser mostrados uno al lado del otro.

Una caja es dibujada para resumir el 50% de nuestro conjunto de datos comenzando en el percentil 25 y finalizando en el percentil 75. La mediana, o percentil 50, es dibujado como una línea. El rango intercuartílico o IQR, es calculado como 1,5 por la diferencia entre el cuartirl 75 y el cuartil 25. Las líneas llamadas bigotes son dibujadas extendiéndose desde el percentil 25 y el 75, este rango muestra observaciones sensibles a dicha distribución.

Observaciones fuera de los bigotes podrían ser outliers y son dibujados como pequeños círculos.

Se trata de un gráfico que nos permite ver la distribución de nuestra muestra de datos como alternativa al histograma. Nos puede ayudar a tener una idea rápida del rango en el cual se mueven los valores comunes y los valores sensibles. Puesto que no estamos viendo la forma de la distribución, este gráfico es usual cuando nuestra muestra de datos presenta una distribución desconocida.

Python dentro de su librería matplotlib dispone de la función **boxplot()** que nos permite visualizar este tipo de gráficos.

```python
# Fijamos la semilla 
seed(1)

#Generamos tres muestras de datos de números aleatorios que provienen de una distribución gaussiana
x = [randn(1000), 5*randn(1000), 10*randn(1000)]

#Graficamos nuestra muestra de datos
plt.figure(figsize = (10,5))
plt.boxplot(x)
plt.show()
```

# Scatter plot o Gráficos de dispersión

Un scatterplot o gráfico de dispersión, es generalmente usado para resumir la relación entre dos pares de muestras. Par de muestras significa dos medidas que son recogidas para una observación dada, tales como el peso y la altura de las personas.

El eje X representa valores observados para la primera muestra, y el eje Y representa valores observados para la segunda muestra.

Los gráficos de dispersión son útiles para mostrar la asociación o correlación entre dos variables.

Python a través de la librería matplotlib dispone de la funcion **scatter()** que nos permite visualizar este tipo de gráficos.

```python
#Fijamos la semilla
seed(1)

#Generamos la primera variable de nuestra muestra de datos
x = 20*randn(1000) + 100

#Generamos la segunda variable de nuestra muestra de datos
y = x + (10 * randn(1000) + 50)

#Generamos nuestro gráfico de dispersion o scatterplot
plt.figure(figsize = (10,5))
plt.scatter(x, y)
plt.show()
```
