```python
from numpy import arange 
import matplotlib.pyplot as plt 
from scipy.stats import norm
import numpy as np
```

# Introducción

Normalmente tenemos la intuición que cuantas más observaciones tengamos mejores resultados obtendremos. Esta intuición subyace a que si recogemos más datos, nuestra muestra de datos será más representativa del dominio del problema.

Existe un teorema en estadística y probabilidad que apoya esta intuición y se trata de un pilar en ambos campos. Este teoremos es conocido como el teoremos de los números grandes y tiene importantes implicaciones en el aprendizaje automático

# Regla de los números grandes

La regla de los números grandes es un teorema de probabilidad y estadística que sugiere que en promedio el resultado de un experimento cuando este se repite múltiples veces se aproximará en mayor medida al valor medio esperado verdadero.

Podemos pensar en una prueba de un experimento como una observación. La repetición de un experimento de forma independiente dará lugar a múltiples pruebas y por lo tanto a múltiples observaciones.

* **Observación:** resultado de una prueba de un experimento.

* **Muestra:** grupo de resultados reunidos de pruebas independientes.

* **Población:** espacio de todas las posibles observaciones.

Haciendo uso de estas definiciones, podemos decir que si el tamaño de la muestra incrementa, el valor medio se aproximará mejor a la media esperada del valor de una población.

# Independientes e idénticamente distribuidos

Es importante tener claro que las observaciones en la muestra deben de ser independientes. Esto significa que la prueba se realiza de una manera idéntica y que no depende de los resultados de cualquier otra prueba.

Esto suele ser razonable y fácil de lograr mediante computación, aunque en ciertas situaciones esto puede llegar a ser algo extremadamente difícil (por ejemplo, ¿cómo lograr que tiradas de un dado sean idénticamente aleatorias?).

En estadística, esta expectativa es llamada muestras independiente e idénticamente distribuidas, o IDD. Esto nos asegura que las muestras son extraídas de una misma distribución de población subyacente.

# Regresión a la media

La regla de los números grandes nos ayuda a comprender el motivo por el cuál no podemos confiar en una sola observación de un experimento aislado.

Se tiende a cometer el error de pensar que el resultado de la media de una pequeña muestra es muy probable que pase, que está cerca de la tendencia central, respecto a la media de la distribución de la población.

Puede que esto no sea cierto, de hecho, lo más seguro es que sea extraño e improbable. La ley nos dice que debemos de repetir el experimento para desarrollar una muestra grande y representativa de observaciones antes de comenzar a realizar inferencias sobre el resultado.

A medida que aumentamos el tamaño de la muestra, la media de la muestra retrocederá hacia la media de la población, es decir, hacia el verdadero valor subyacente. Esto se le conoce como regresión a la media o reversión a la media. Es por esta razón por la que debemos ser escépticos acerca de las inferencias a partir de tamaños de muestras pequeñas, llamadas n pequeñas.

# Ley de los números realmente grandes

La auténtica ley de los números realmente grandes subyace de la regresión a la media. Esta ley nos dice que cuando trabajamos con muestra altamente grandes, aumentamos la probabilidad de observar cosas extrañas. 

Al tener un mayor número de muestras de la población subyacente, las muestras contrandrán eventos raros. Una vez más, debemos de ser cautelosos de no hacer ningún tipo de inferencia a partir de casos aislados.

# Trabajando con un ejemplo

Podemos demostrar la ley de los números grandes a partir de un ejemplo. En primer lugar vamos a proceder a crear una distribución subyacente ideal. Vamos a crear un distribución Gaussiana con media 50 y desviación estándar 5.

```python
#FIjamos el eje x
x = arange(30,70, 1)

#Fijamos el eje y 
y = norm.pdf(x, 50, 5)

#Mostramos el resultado 
plt.figure(figsize = (10,5))
plt.plot(x, y)
plt.show()
```

A continuación vamos a proceder a extraer muestras independientes de la muestra original. Para cada uno de estas muestras vamos a proceder a calcular la media. Esperamos, debido a la regla de los números grandes, a que si incrementamos el tamaño de nuestra muestra su media se aproximará en mayor medida a la media de la población original. 

```python
#Fijamos la semilla
np.random.seed(1)

#Nos creamos una lista con los distintos tamaños de la muestra
sizes = [x for x in arange(10, 20000, 200)]

#Extraemos muestras y calculamos media
result = [np.mean(5*np.random.randn(size) + 50) for size in sizes]

#Vemos el resultado 
plt.figure(figsize = (10,5))
plt.scatter(sizes, np.array(result) - 50)
plt.show()
```

Podemos ver que a medida que aumenta la muestra el resultado suele converger al valor de la media original.
