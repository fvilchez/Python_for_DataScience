```python
from numpy import arange
from scipy.stats import norm, t, chi2
import matplotlib.pyplot as plt
```

# Introducción 

Una muestra de datos forma un función de distribución, la distribución más conocida es la distribución gaussianna también conocida como distribución normal. La función de distribución nos proporciona una función matemática parametrizada que puede ser usada para calcular la probabilidad de cualquier individuo perteneciente a dicha distribución. Esta distribución describe la agrupación o la densidad de las observaciones, llamada función de densidad de probabilidad. También podemos calcular la probabilidad de que una observación tenga un valor igual o menor a un valor dado.

# Distribuciones

Podemos pensar en una función de distribución como una función que describe la relación entre observaciones en un espacio muestral. Muchos datos se ajustan a funciones matemáticas bien conocidas, tales como la distribución gaussianna. Una vez una función de distribución es conocida, esta puede usarse para calcular relaciones cuantitativas, tales como probabilidad de observaciones.

# Funciones de densidad

Las distribuciones son a menudo descritas en términos de densidad o funciones de densidad. Las funciones de densidad son funciones que describen como cambia la proporción de los datos o la probabilidad en el rango de la distribucion. Tenemos dos tipos de funciones de densidad:

* **Función densidad de probabilidad:** calcula la probabilidad de observar un valor dado.

* **Función de densidad acumulativa:** calcula la probabilidad de que una observación sea igual o menor a un valor.

Una función de densidad de probabilidad, o PDF, puede ser usada para calcular la probabilidad de una observación dada una distribución. Puede ser usada para sumarizar la probabilidad de observaciones a través de un espacio muestral de la distribución. 

Un función de densidad acumulada, o CDF, es una forma diferente de pensar sobre la probabilidad de observar valores. La CDF calcula la probabilidad acumulada para una observación y todas las observaciones anteriores en el espacio muestral. 

Tanto la PDF como la CDF son funciones continuas. El equivalente a la PDF para una distribución discreta es la PMF.

# Distribución Gaussiana

La distribución Gaussianna, llamada así por su descubridor Carl Friedrich Gauss, es el foco de estudio de mucho de los campos de la estadística. Esto se debe a que los datos de muchos campos de estudio pueden ser descritos haciendo uso de una distribución Gaussiana, por lo que esta distribución es a menudo llamada normal, debido a que es muy común. Una distribución Gaussiana puede ser descrita usando dos parámetros:

* **media:** denotada por la letra griega mu, es el valor esperado de la distribución.

* **varianza:** se denota por la letra sigma elevada al cuadrado, describe como de concentrados están los datos entorno a la media

Es muy común usar el calculo normalizada de la varianza conocido como desviación estándar:

* **desviación estándar:** denotada por sigma, describe la concentración normalizada de las observaciones respecto a la media.

Podemos trabajar con la distribución Gaussiana mediante la librería Scipy, a partir de su modulo **norm**. La función **norm.pdf()** puede ser usada para crear una distribución gaussiana, a partir de un espacio muestral, una media y una desviación estándar.

```python
#Nos creamos nuestra espacion muestral
sample_space = arange(-10, 10, 0.001)

#Fijamos la media y la desviación estándar
media = 0.0
desv_std = 1.0

#Calculamos la pdf
pdf = norm.pdf(sample_space, media, desv_std)

#Ploteamos el resultado
plt.plot(sample_space, pdf)
plt.show()
```

Este gráfico muestra en el eje x cada uno de los elementos que componen nuestro espacio muestral y en el eje y la probabilidad de que se de ese elemento. 

También podemos hacer uso de la función **norm.cdf()** para crear y representar la función de densidad acumulada.

```python
#Nos creamos la función de densidad acumulada
cdf = norm.cdf(sample_space)

#Vemos el resultado 
plt.plot(sample_space, cdf)
plt.show()
```

En este caso podemos ver en el eje x cada uno de los valores de que toma nuestra espacio muestral y en el eje y la probabilidad acumulada. Podemos observar como par ael caso de 2 por ejemplo, tenemos una probabilidad acumulada de aproximadamente el 90%, es decir, la probabilidad de obtengamos un valor por debajo de 2 es bastante probable, mientras que obtener un valor por encima es poco probable.

# Student's t-Distribution

La distribución de Student's o t-distribution, es llamada así por el pseudónimo Student de William Sealy Gosset. Esta es una distribución que surge cuando intentamos estimar la media de una distribución normal con diferentes tamaños muestrales.

Como tal, es útil cuando se describe la incertidumbre o error relacionado con la estimación estadística de una población que ha sida extraida de una distribución Gaussiana.

Aunque esta distribución no puede ser usada directamente, podemos estimar valores requeridos para una distribución como parámetros en otros métodos estadísticos. La distribución puede ser descrita mediante un único parámetro:

* **número de grados de libertad:** se denotan por la letra griega nu, indican el número de grados de libertad.

La clave para poder hacer uso de la distribución t es conocer el número de grados de libertad deseados. El número de grados de libertad describe el número de piezas de información usadas para describir la población. Por ejemplo, la media tiene n grados de libertad puesto que todas las observaciones en la muestra son usadas para estimar la media de la población. Una cantidad estadística que hace uso de otra cantidad estadística en su cálculo, se le debe restar 1 grado de libertad. Por ejemplo, a la hora de calcular la varianza hacemos uso de la media estadística, por lo tanto la varianza tiene un total de n-1 grados de libertad.

Scipy nos proporciona herramientas para poder trabajar con la distribución t, para esto disponemos del modulo **stats.t**. La función **t.pdf()** puede ser usada para crear una distribución de tipo t Student's con un número específico de grados de libertad.

```python
#Definimos las distribución de parámetros 
sample_space = arange(-10, 10, 0.001)

#Incamos los grados de libertad
dof = len(sample_space) - 1

#Calculamos la distribución
t_pdf = t.pdf(sample_space, dof)

#Visualizamos el resultado
plt.plot(sample_space, t_pdf)
plt.show()
```

A simple ojo parece ser exactamente igual a la distribución gaussiana, sin embargo, se puede observar que existe una mayor probabilidad de obtener valores en los extremos de las colas.

También podemos crear la función de distribución acumulada, a partir de la función **t.cdf()**.

```python
#Nos creamos la función acumulada
t_cdf = t.cdf(sample_space, dof)

#Vemos los resultados
plt.plot(sample_space, t_cdf)
plt.show()
```

# Distribución Chi-Cuadrado

Como la distribución de Student's, la distribución Chi-Cuadrado es también usada en métodos estadísticos donde los datos son extraidos de una distribución Gaussiana para cuantificar la incertidumbre. Por ejemplo, la distribución Chi-Cuadrado es usada en el test estadístico Chi-Cuadrado para independencia. La distribución Chi-Cuadrado tiene un único parámetro:

* **Grados de Libertad:** es denotado por k

Una observación en una distribución Chi-Cuadrado es calculada como la suma de las k observaciones al cuadrado, estas k bservaciones son extraídas de una distribución Gaussiana. Al igual que en la distribución de Student's los datos no se ajustan a una distribución Chi-Cuadrado, sin embargo, las observaciones son extraídas de esta distribución en el cálculo de métodos estadísticos para una muestra de datos Gaussianos.

Scipy proporciona el módulo **stats.chi2** para realizar cálculos estadísticos para la distribución Chi-Cuadrado. La función **chi2.pdf()** puede ser usada para calcular la distribución Chi-Cuadrado para un espacio muestral entre 0 y 50 y con 20 grados de libertad. 

```python
#Nos creamos el espacio muestral
sample_space = arange(0, 50, 0.001)

#Indicamos los grados de liberta 
dof = 20

#Calculamos la pdf
pdf = chi2.pdf(sample_space, dof)

#Vemos el resultado 
plt.plot(sample_space, pdf)
plt.show()
```

Podemos ver como el valor esperado es aproximadamente 20. Esto es intuitivo, si pensamos que para una distribución gaussiana la mayoría de los valores se encuentran entre -1 y 1 y puesto que la suma de las muestras aleatorias cuadradas del estándar Gaussiana sumaría un poco menos del número de grados de libertad. Aunque la distribución tiene forma de campana podemos ver que esta no es simétrica.

La función **chi2.cdf()** nos permite calcular la función de densidad acumulada a través de un espacio muestral.

```python
#Nos creamos la cdf
cdf = chi2.cdf(sample_space, dof)

#Vemos el resultado 
plt.plot(sample_space, cdf)
plt.show()
```
