# Subset time series data

Cuando estamos trabajando con series temporales, es posible que queramos visualizar únicamente un subconjunto de nuesta serie. Pandas nos permite realizar esta selección de forma sencilla.

```python
#Seleccionamos un subconjunto de nuestros datos entre 1945 y 1950
discoveries_subset_1 = discoveries.loc['1945' : '1950']

#Nos creamos nuestro plot
ax = discoveries_subset_1.plot(color = 'blue', fontsize = 16)

#Vemos el resultado
plt.show()
```

# Add vertical and horizontal markers

Matplotlib nos permite realizar anotaciones sobre nuestro plot, esto nos puede ayudar a hacer enfásis en determinadas zonas de nuestra serie temporal. Para ello diponemos de **axvline** para líneas verticales y **axhline** para líneas horizontales.

```python
#Nos creamos nuestro gráfico
ax = discoveries.plot(color = 'blue', fontsize = 6)

#Nos creamos nuestra línea vertical
ax.axvline('1939-01-01', color = 'red', linestyle = '--')

#Nos creamos nuestra línea horizontal
ax.axhline(4, color = 'green', linestyle = '--')

#Vemos el resultado 
plt.show()
```

# Add shaded regions to your plot

Es posible que no solo queramos señalar un punto específico, sino que queramos remarcar una determinada zona de nuesta serie temporal.

```python
ax = discoveries.plot(color = 'blue', fontsize = 6)

#Generamos la región de interés sombreada
ax.axvspan('1900-01-01', '1915-01-01', color = 'red', alpha = 0.3)
ax.axhspan(6, 8, color = 'green', alpha = 0.3)

#Vemos el resultado
plt.show()
```
