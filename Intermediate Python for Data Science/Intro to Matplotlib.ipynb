## Line plot(II)

A continuación vamos a proceder a cargar el conjunto de datos gapminder, este conjunto de datos tiene información respecto a la población, la esperanza de vida y la renta per capita en los distintos países en el mundo.

```python
#Realizamos la importación de los datos
import pandas as pd
gm = pd.read_csv('gapminder.csv')
```

```python
#Extraemos la información de las columnas de la renta per capita ,la esperanza de vida y población
gdp_cap = list(gm.loc[:, 'gdp_cap'])
life_exp = list(gm.loc[:, 'life_exp'])
pop = list(gm.loc[:, 'population'])
```

Existen situaciones donde los gráficos de línea no son una buena opción para representar nuestros datos.

```python
#Nos creamos nuestro gráfico tipo línea
_ = plt.plot(gdp_cap, life_exp)
#Mostramos el resultado por pantalla
plt.show()
```

Los gráficos de tipo línea se recomiendan usar cuando la variable independiente es de tipo temporal.

## Scatter plot (I)

Existen situaciones en las cuales nos interese observar si nuestras variables están correladas, para esto los gráficos de tipo scatter plot son los más recomendables.

```python
#Nos creamos nuestro gráfico de tipo scatter
plt.scatter(gdp_cap, life_exp)
#Mostramos el resultado por pantalla
plt.show()
```

En este caso se puede ver de forma bastante clara que los países con una mayor esperanza de vida son aquellos en los que su renta per capita es mayor.

## Scatter plot (II)

Los Scatter plot también nos puede arrojar información como la decorrelación entre dos variables.

```python
#Nos creamos nuestro gráfico de tipo scatter
_ = plt.scatter(pop, life_exp)
#Mostramos los resultados por pantalla
plt.show()
```

En este caso podemos ver que no existe correlación entre la población de un país y la esperanza de vida.

## Build a histogram (I)

Los histogramas se tratan de un tipo de gráfico que nos permite visualizar la distribución de nuestro conjunto de datos. A la hora de realizar este tipo de gráficos Python dispone de la función plt.hist(), donde el primer parámetro que le pasamos es nuestro conjunto de datos.

```python
#Nos creamos nuestro primer histograma
plt.hist(life_exp, edgecolor='black')
#Mostramos el por pantalla
plt.show()
```

## Build a histogram (II): bins

En el histograma anterior no hemos especificado el número de bins, en el caso que no indiquemos este parámetro, Python por defecto le da el valor de 10. El número de bins en un histograma es muy importante. Seleccionar un número elevado de bins puede hacer que pasemos por alto ciertos detalles, mientras que si mostramos seleccionamos un elevado número de bins puede que compliquemos en exceso. Para seleccionar el número de bins, Python dispone del parámetro bins.

```python
#Nos creamos el mismo histograma anterior pero con 20 bins
plt.hist(life_exp, bins = 20, edgecolor = 'black')
#Mostramos el resultado por pantalla
```

## Labels

Una vez tenemos nuestro gráfico representado, llego el momento de mejorarlo, para ello podemos indicar que representan cada uno de los ejes y dar una descripción del gráfico poniéndolo un título.

```python
#Nos creamos un plot básico de tipo scatter
plt.scatter(gdp_cap, life_exp, edgecolor = 'black')
plt.xscale('log')

#Le agregamos los nombres a los ejes x e y
plt.xlabel('GDP per capita [in USD]')
plt.ylabel('Life Expectancy [in years]')

#Le agragamos un título
plt.title('World Development')

#Mostramos el resultado por pantalla
plt.show()
```

## Ticks

Los ticks nos permite clarificar aún mejor los ejes

```python
#Nos creamos un plot básico de tipo scatter
plt.scatter(gdp_cap, life_exp, edgecolor = 'black')
plt.xscale('log')

#Le agregamos los nombres a los ejes x e y
plt.xlabel('GDP per capita [in USD]')
plt.ylabel('Life Expectancy [in years]')

#Le agragamos un título
plt.title('World Development')

#Añadimso ticks al plot anterior
plt.xticks([1000, 10000], ['1k', '10k'])

#Mostramos el resultado por pantalla
plt.show()
```

## Sizes

Debido a que todos los scatters parecen indistinguibles unos de otros, vamos a proceder a poner el tamaño de estos en función de la población de cada uno de los países.

```python
#Nos creamos un numpy array de la población
import numpy as np
np_pop = np.array(pop)
np_pop = np_pop/1000000

#Nos creamos un plot básico de tipo scatter y decimos que el tamaño sea por población
plt.scatter(gdp_cap, life_exp, s = np_pop, edgecolor = 'black')
plt.xscale('log')

#Le agregamos los nombres a los ejes x e y
plt.xlabel('GDP per capita [in USD]')
plt.ylabel('Life Expectancy [in years]')

#Le agragamos un título
plt.title('World Development')

#Añadimso ticks al plot anterior
plt.xticks([1000, 10000], ['1k', '10k'])

#Mostramos el resultado por pantalla
plt.show()
```

## Colors

También podemos modificar los colores, por ejemplos supongamos que queremos que cada continente sea de un color diferente.

```python
#Nos creamos el diccionario que mapea los continentes
dict = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'Americas':'yellow',
    'Oceania':'black'
}

#Mapeamos el dataframe de forma adecuada
gm = gm.replace({"cont": dict})
#Extraemos en una lista la información del color
color = list(gm.loc[:,'cont'])
```

```python
#Nos creamos un numpy array de la población
import numpy as np
np_pop = np.array(pop)
np_pop = np_pop/1000000

#Nos creamos un plot básico de tipo scatter y decimos que el tamaño sea por población y además agregamos colores
plt.scatter(gdp_cap, life_exp, s = np_pop, edgecolor = 'black', c = color, alpha = 0.8)
plt.xscale('log')

#Le agregamos los nombres a los ejes x e y
plt.xlabel('GDP per capita [in USD]')
plt.ylabel('Life Expectancy [in years]')

#Le agragamos un título
plt.title('World Development')

#Añadimso ticks al plot anterior
plt.xticks([1000, 10000], ['1k', '10k'])

#Añadimos grid
plt.grid(True)

#Mostramos el resultado por pantalla
plt.show()
```
