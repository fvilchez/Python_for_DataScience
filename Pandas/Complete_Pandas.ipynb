# Tipos de Objetos Pandas

En su nivel más básico, los Pandas objetos pueden considerarse como una versión mejorada de las matrices estructuradas de **Numpy**, en las que las filas y columnas se identifican mediante etiquetas en lugar de números enteros. **Pandas** proporciona una serie de herramientas, métodos y funcionalidades extramademente útiles para el análisis de datos, pero antes se debe comprender como se estructura **Pandas**. ## Pandas Series

Un pandas series se trata de un array unidimensional indexada. Este tipo de datos pueden ser creados a partir de un array unidimensional o de una lista de la siguiente forma.

```python
#Importamos las librerías pandas y numpy
import pandas as pd
import numpy as np

#Nos creamos un objeto tipo serie
data = pd.Series([0.25, 0.5, 0.75, 1.0])

#Mostramos el resultado
print(data)
```

Podemos ver que nuestra salida dispone tanto de una secuencia de valores(los datos de nuestra lista) y una secuencia de índices. Podemos acceder tanto a los valores como a los índices mediante los atributos: **values** y **index** respectivamente.

```python
#Accedemos a los valores de nuestra serie
print(data.values)

#Vemos el tipo 
print(type(data.values))
```

Podemos ver como los valores no son más que una simple array unidemensional.

```python
#Accedemos a los índices
print(data.index)

#Accedemos al tipo de dato
print(type(data.index))
```

Los índices podemos ver que son de tipo **pd.index**, este tipo de dato será discutido más adelante.

Podemos acceder a los datos a través de sus índices, de igual forma que hacíamos en un numpy array, haciendo uso de los corchetes.

```python
#Accedemos al primer valor de nuestra serie
print(data[0])
```

```python
#Accedemos al segundo y al tercer elemento de nuestra serie
print(data[1:3])
```

Por lo que hemos visto hasta ahora un Pandas Series y un array unidimensional son prácticamente iguales, con la diferencia de que nuestro objeto pandas series tiene un índice, mientra un Numpy array tiene por defecto el valor entero como índice, nuestro objeto pandas series nos permite modificar dicho índice, es decir, el índice no tiene por que ser un valor entero. Para modificar el índice de un objeto pandas series disponemos dentro de la función **pd.Series()** del argumento **index**.

```python
#Nos creamos un objeto pandas series modificando los índices
data = pd.Series([0.25, 0.5, 0.75, 1.0], index = ['a', 'b', 'c', 'd'])

#Mostramos el resultado
print(data)
```

Una vez tenemos nuestros índice asignados, igual que hemos visto anteriormente podemos acceder a cada uno de los elementos a partir de los valores de nuestros nuevos índices.

```python
#Accedemos a patir de nuestros nuevos índices
print(data['b'])
print(data['a':'c'])
```

Tras esto podríamos pensar que un objeto pandas series se trata de una especialización de un diccionario. Un diccionario se trata de una estructura que asigna un conjunto de clave arbitrarias a un conjunto de valores arbitrarios, mientras que una **Serie** asigna un conjunto de claves tipadas a un conjunto de valores tipados. Esta tipificación es importante: de igual forma que el código compilado que corre detrás de un Numpy array la hace más eficiente que una lista Python para ciertas operaciones, el tipado de los datos de una Serie hace que esta sea mucho más eficiente que un diccionario para ciertas operaciones.

Podemos crear series a partir de diccionarios y diccionarios a partir de series.

```python
#Nos creamos una serie a partir de un diccionario
champions_dict = {'Real_Madrid': 13, 'Milan': 7, 'Liverpool': 5, 'Bayern_Munich': 5}

#Nos creamos nuestro pandas series a partir del diccionario
champions_series = pd.Series(champions_dict)

#Mostramos el resultado
print(champions_series)
```

Por defecto las series son ordenadas por clave en sentido ascendente. 

También podemos crear series de valores constantes, es decir, que toda la serie contenga el mismo valor.

```python
#Nos creamos un arrray de valor constante
constant_serie = pd.Series(5, index = [100, 200, 300])

#Mostramos el resultado
print(constant_serie)
```

A partir del índice podemos especificar que conjunto de datos queremos agregar a nuestra serie.

```python
#Nos creamos una serie especificando los índices que deseamos agregar
serie = pd.Series({2:'a', 1:'b', 3:'c'}, index = [1,2])

#Mostramos el resultado
print(serie)
```

## Objetos Pandas DataFrame

Si una serie es una analogía de un array unidimensional con índices flexibles, un **DataFrame** es análogo a un array bidimensional con flexibilidad de indexación tanto para filas y columnas, podemos pensar en un **DataFrame** como objetos de tipo serie alineados. Aquí la palabra alineado quiere decir que comparten el mismo índice.

```python
#Nos creamos nuestra primera serie
champions_dict = {'Real_Madrid': 13, 'Milan': 7, 'Liverpool': 5, 'Bayern_Munich': 5}
champions_series = pd.Series(champions_dict)

#Nos creamos nuestra segunda serie
liga_dict = {'Real_Madrid': 33, 'Milan': 18, 'Liverpool': 18, 'Bayern_Munich': 28}
liga_series = pd.Series(liga_dict)

#Nos creamos nuestro pandas DataFrame
info_teams = pd.DataFrame({'champions_league': champions_series, 'ligas': liga_series})

#Mostramos el resultado
print(info_teams)
```

Al igual que las Series tienen el atributo **index** para acceder a los índices, nuestros objetos Pandas DataFrame también disponen de este atributo.

```python
#Mostramos los índices
print(info_teams.index)
```

Los objetas Pandas DataFrame también disponen del atributo **columns** que nos permite obtener el nombre de las columnas de nuestro DataFrame. Este atributo nos retorna un objeto de tipo index.

```python
#Obtenemos y mostramos los valores de las columnas
columns = info_teams.columns
print(columns)
```

De forma similar podemos pensar que un **DataFrame** es un diccionario especializado. Mientras los diccionarios mapean claves con valores, un **DataFrame** mapea columnas con Series. Por ejemplo, si intentamos acceder a las champions_league ganadas por cada equipo, veremos que esto nos retorna un objeto tipo Pandas Series.

```python
champions_leagues = info_teams['champions_league']
print(champions_leagues)
print(type(champions_leagues))
```

Aquí vamos a destacar un punto de confusión en la gran mayoría de las personas: mientras que en **Numpy Array** si hacemos uso de la notación **data[0]** esto nos retorna el valor de la **primera fila**, en el caso de un **Pandas DataFrame** la notación **data['col0']** nos retorna el valor de la **primera columna**. 

### Construyendo objetos Pandas Dataframe

Los objetos Pandas DataFrame se pueden construir de diversas formas, a continuación veremos algunas de estas.

#### A partir de un Numpy Array 2-dimensional

## Indexado: loc, iloc y ix

Visto lo anterior seleccionar elementos haciendo uso del indexado convencional puede resultar confuso. Por ejemplo, si nuestra serie esta indexada explicitamente por índices enteros, si hacemos uso de **data[1]** estamos accediendo al índice específico 1. Sin embargo si hacemos uso de **data[1:3]** estamos accediendo al indexado convencional de Python.

```python
#Nos creamos nuestra serie
data = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
print(data)
```

```python
#Accedemos al elemento con el índice 1
print(data[1])
```

```python
#Accedemos a los elementos por indexación de Python
print(data[1:3])
```

Debido a esta potencial confusión en el caso de indexado de índices enteros, Pandas nos proporciona algunos atributos especiales de indexado.

### loc

El primero de estos atributos es **loc**, este atributo nos permite indexar siempre mediante índice.

```python
#Hacemos uso de iloc para indexar por índice
print(data.loc[1])
```

```python
#Hacemos uso de iloc para indexar por índice
print(data.loc[1:3])
```

### iloc

El segundo de estos atributos es **iloc**, este atributo nos permite indexar siempre en el estilo de indexación de Python convencional.

```python
#Indexamos con iloc
print(data.iloc[1])
```

```python
#Indexamos múltiples elementos con iloc
print(data.iloc[1:3])
```

### ix

Se trata de un híbrido de los dos anteriores, y para objetos de tipo series es equivalente al indexado estándar con **[]**. Este forma de indexar cobra mucho más sentido cuando hacemos uso de DataFrames.

## Selección de datos en DataFrames

Debemos recordar que los dataframes no son más que arrays bidimensionales o arrays estructuradas, o diccionarios de series.

### DataFrame como diccionario

La primera analogía que podemos considerar es que un dataframe es un diccionario que relaciona objetos de tipo Series. 

```python
#Nos generamos nuestro DataFrame con un diccionario de series
area = pd.Series({'California': 423967, 'Texas': 695662,
'New York': 141297, 'Florida': 170312,
'Illinois': 149995})
pop = pd.Series({'California': 38332521, 'Texas': 26448193,
'New York': 19651127, 'Florida': 19552860,
'Illinois': 12882135})

data = pd.DataFrame({'area' : area, 'pop' : pop})

#Mostramos el resultado 
print(data)
```

Para acceder a alguna de las series, podemos acceder haciendo uso del indexado de tipo diccionario.

```python
#Accedemos a la Serie area
print(data['area'])
```

```python
#También podemos acceder haciendo uso del estilo atributo
print(data.area)
```

Debemos tener en mente que esta forma de acceder a la serie (mediante estilo atrbuto) no es es siempre posible. Por ejemplo si los nombres de las columnas no son de tipo string o si los nombres de las columnas entran en conflicto con métodos asociados a un DataFrame. Por ejemplo, nuestro DataFrame tiene un método **pop()** asociado por lo que no podremos acceder a la columna pop haciendo uso de la notición (data.pop).

Podemos hacer uso del estilo diccionario para agregar una nueva columna.

```python
#Sin embargo si le pasamos un indexado por columna accedemos a la columna
print(data['area'])
```

# Operando con datos en Pandas

Unas de las piezas esenciales de Numpy es la abilidad de realizar de forma rápida operaciones simples (suma, resta, multiplicación, división, etc) y también operaciones más sofisticadas (operaciones trigonómetricas, exponenciales, logarítmicas, etc). Pandas también dispone de muchas de estas funcionalidades de Numpy, y las funciones universales son la clave de todo esto.

## Ufuncs: preservando el índice

Debido a que Pandas está diseñado para trabajar con Numpy, cualquier función universal de Numpy, podrá ser usada en objetos de tipo Pandas Serie o Pandas DataFrame. Vamos a comenzar definiendo una simple serie y dataframe.

```python
import pandas as pd
import numpy as np

#FIjamos la semilla
rng = np.random.RandomState(42)


#Nos creamos un objeto de tipo serie
ser = pd.Series(rng.randint(0, 10, 4))
print(ser)
```

```python
#Nos creamos un dataframe
df = pd.DataFrame(rng.randint(0,10, (3,4)), columns = ['A', 'B', 'C', 'D'])
print(df)
```

Si aplicamos a nuestra serie o dataframe funciones universales, podemos ver como los objetos serie y dataframe preservan el índice.

```python
#Aplicamos la función exp a nuestra serie
np.exp(ser)
```

```python
#Aplicamos la función sin a cada uno de los elementos de nuestro dataframe
np.sin(df * np.pi / 4)
```

## Ufuncs: alineamiento de índices

Para operaciones binarias entre dos objetos de tipo serie o dataframe, Pandas alinea índices en el proceso de realizar la operación. Esto es muy conveniente cuando estamos trabajando con datos incompletos.

### Alineamiento de índices en series

Supongamos que estamos combinando dos fuentes de datos, una serie dispone del top 3 de los estados de USA con mayor población y otra serie dispone del top 3 de los estados USA con mayor extensión.

```python
#Nos creamos las dos series
area = pd.Series({'Alaska': 1723337, 'Texas': 695662,'California': 423967}, name='area')
population = pd.Series({'California': 38332521, 'Texas': 26448193,'New York': 19651127}, name='population')

#Dividimos ambas series con el objetivo de tener el número de habitantes por km2
population / area
```

EL resultado contiene la unión de los índices de las dos series de entrada, estos índices se podrían calcular haciendo uso de la operación aritmética:

```python
#Vemos la unión de los índices de nuestras series
area.index | population.index
```

Cualquier índice de la operación que no se encuentra en alguna de las series será marcado como NaN (Not a Number), esta es la forma en la cual Pandas trata a los valores perdidos. Esta forma de machear índices es usada por Python en cualquier operación aritmética y cualquier valor faltante es marcado como NaN.

```python
#Nos creamos dos series 
A = pd.Series([2, 4, 6], index=[0, 1, 2])
B = pd.Series([1, 3, 5], index=[1, 2, 3])

#Hacemos la suma
A + B
```

Si este relleno como valores NaN no es lo que deseamos, podemos modificar la forma de rellenar los valores faltantes haciendo uso de los métodos apropiados. Por ejemplo si hacemos uso de A.add(B) que es lo mismo que A + B, podremos hacer uso del parámetro **fill_value** en el cual indicamos con que valor queremos que nos rellene los valores faltantes.

```python
#Hacemos la suma con add y hacemos uso del parámetro fill_value
A.add(B, fill_value = 0)
```

### Alineamiento de índices en un dataframe

Un alineamiento similar por columna e índice se produce en los DataFrame cuando realiamos operaciones entre dataframes.

```python
A = pd.DataFrame(rng.randint(0, 20, (2, 2)), columns=list('AB'))
print(A)
```

```python
B = pd.DataFrame(rng.randint(0, 10, (3,3)), columns = list('ABC'))
print(B)
```

```python
#Realizamos la suma entre ambos dataframes
A + B
```

Podemos fijarnos como los índices son alineados de forma correcta, independientemente del orden en los dos objetos y los índices en el resultados están ordenados. También podemos hacer uso de las funciones universales, para hacer uso del argumento **fill_value**.

```python
#Calculamos la media del dataframe A
value = A.stack().mean()

#Hacemos la suma y rellenamos
A.add(B, fill_value = value)
```

# Funciones universales: Operaciones entre DataFrames y Series

Las operaciones entre un DataFrame y una Serie son similares a las operaciones entre arrays bi-dimensionales y uni-dimensionales. Consideremos la operación donde intentamos encontrar la diferencia entre un array bi-dimensional y una de sus filas.

```python
#Hacemos la diferencia entre de dicho DataFrame con su primera fila
df - df.loc[2]
```

Si lo que queremos es realizar la operación por columna debemos de hacer uso de las funciones vistas anteriormente que nos permite modificar el parámetro **axis**.

# Manejo de datos perdidos

La gran diferencia entre los datos encontrados en muchos tutoriales y los datos que nos encontramos en la vida real, es que los datos que nos encontramos en la vida real raramente están limpios y homogeneos. En general, la gran mayoría de conjunto de datos interesantes tienen una gran cantidad de valores perdidos. Además el uso de diferentes fuentes de datos suelen dar lugar a datos faltantes de distintas formas. A continuación vamos a discutir algunas consideraciones generales a la hora de tratar con valores perdidos, como Pandas representa estos valores, y veremos las herramientas de las que dispone Pandas para hacer el tratamiento de estos valores perdidos. Nos referiremos a valores faltantes mediante **null**, **NaN** o **NA**.

La forma en que Pandas trata los valores perdidos se ve limitada por su dependencia con el paquete Numpy, que no tiene una noción de incorporar valores perdidos para tipos de datos que no son de punto flotante. Para esto Python hace uso de dos tipos de valores faltantes: los valores faltantes de punto flotante **NaN** y lo objetos tipo **None**.

## None: Valores perdidos en Python

El primer valor usado por Pandas es  **None**. Puesto que **None** se trata de un tipo de objeto en Python, este no puede ser usado de forma arbitraria en cualquier Pandas/Numpy array, solo en las matrices con tipos de dato object.

```python
#Nos creamos un numpy array con uno de sus valores a None
vals1 = np.array([1, 2, 3, None])
vals1
```

Si bien este tipo de dato en ciertas ocasiones puede ser beneficioso, en otros muchos casos no es la forma más adecuada a la hora de operar con Python.

```python
for dtype in ['object', 'int']:
    print('dtype = ' , dtype)
    %timeit np.arange(1E6, dtype = dtype).sum()
```

Si hacemos operaciones de agregación como **sum()** o **mean()** en un array que contiene valores **None**, esto nos generará un error.

```python
#Hacemos una suma de valores
vals1.sum()
```

Esto refleja que la suma entre un valor de tipo **None** y uno tipo **int** no está definido.

## NaN: valores perdidos en datos numéricos

La otra forma de representar los valores perdidos, **NaN** (Not a Number) es diferente. Este se trata de un valor especial que es reconocido por todos los sistemas que usan el estándar en coma flotante del IEEE.

```python
vals2 = np.array([1, 2, 3, np.nan])
vals2.dtype
```

Podemos ver que Numpy selecciona el tipo de dato **float64** para esta array: esto significa que sobre esta array podemos hacer uso de todas las operaciones rápidas que corren bajo código compilado. Debemos pensar en **NaN** como un virus, es decir, infecta todo aquello que toca. De forma que cualquier operación que contenga un valor **NaN** devolverá **NaN**.

```python
#Le sumamos 1 a NaN
print(1 + np.nan)
```

```python
#Hacemos una agregación 
print(vals2.sum())
```

Numpy dispone de agregaciones especiales que ignoran estos valores perdidos:

* np.nansum()

* np.nanmin()

* np.nanmax()

```python
#Hacemos uso de las funciones que ignoran los valores perdidos
print(np.nansum(vals2), np.nanmin(vals2), np.nanmax(vals2))
```

Debemos tener en mente que **NaN** es un valor exclusivo para datos de tipo coma flotante, no es equivalente al valor **NaN** para enteros, strings o otros tipos.

# Operando con valores perdidos

Como ya hemos visto, Pandas trata los valores **None** y **NaN** indistintamente para indicar valores perdidos. Para facilitar el tratamiento de estos valores existen métodos que nos permiten detectar, imputar o eliminar estos valores perdidos, estos métodos son:

* **isnull():** genera una máscara booleana indicando los valores perdidos.

* **notnull():** opuesto a isnull(), es decir, genera una máscara indicando los valores que no son perdidos.

* **dropna():** retorna una versión filtrada de los datos.

* **fillna():** retonar una versión de los datos con los valores perdidos imputados o rellenos.

### Detectando valores perdidos

Las estructuras de datos tipo Pandas, tiene dos métodos muuy útiles a la hora de detectar valores perdidos: **isnull()** y **notnull()**. Cada una de estas retorna una máscara sobre los datos

```python
#Nos creamos un objeto tipo Series
data = pd.Series([1, np.nan, 'hello', None])

#Aplicamos el método isnull()
data.isnull()
```

Como podemos ver esto nos retorna una máscara con el valor de ***True** en aquellos índices en los que tenemos valores perdidos.

```python
#Aplicamos el método notnull()
data.notnull()
```

En este caso tenemos una máscara donde toma el valor **True** para aquellos índices en los que no tenemos valores perdidos. Podemos hacer uso de esta máscara para seleccionar aquellos valores que no son perdidos.

```python
#Seleccionamos aquellos valores que no son perdidos
data[data.notnull()]
```

Esto dos método producen un resultado similar cuando lo aplicamos sobre objetos de tipo DataFrame.

### Eliminando valores perdidos

Además del uso de la máscara, existen otros métodos para eliminar valores perdidos que son más convenientes de usar:, **dropna()** (elimina las observaciones con valores perdidos) y **fillna()** (rellena los valores perdidos).

```python
#Hacemos uso del método dropna()
data.dropna()
```

A la hora de aplicar este método en un DataFrame existen más opciones. Consideremos el siguiente DataFrame

```python
#Nos creamos nuestro DataFrame
df = pd.DataFrame([[1, np.nan, 2], [2, 3, 5], [np.nan, 4, 6]])
print(df)
```

No podemos eliminar valores perdidos sueltos, solo podemos eliminar filas o columnas completas. Dependiendo de la situación podremos realizar una u otra.

Por defecta la función **dropna()** elimina filas

```python
#Aplicamos el método dropna sobre nuestro DataFrame
df.dropna()
```

Podemos especificar que elimine columnas haciendo uso del parámetro opcional **axis**

```python
#Eliminamos columnas
df.dropna(axis = 'columns')
```

Podríamos por ejemplo querer eliminar aquellas filas o columnas que tuviesen todos sus valores como valores perdidos, o aquellas que tenga un determinado porcentaje de valores perdidos. Esto se puede hacer mediante los parámetros **how**  y **thresh** que nos permite afinar el número de valores perdidos que deseamos eliminar.

Por defecto el parámetro **how** toma el valor de **any**, lo que significa que cualquier fila o columna dependiendo del valor del parámetro **axis** que contenga algún valor perdidos será eliminada. Si a este parámetro le especificamos el valor **all** eliminará solo aquelllas filas o columnas en la que todos sus valores sean valores perdidos.

```python
#Nos creamos una nueva columna que contiene todos sus valores perdidos
df[3] = np.nan
print(df)
```

```python
#APlicamos el método dropna() con how = all
df.dropna(how = 'all', axis = 'columns')
```

El parámetro **thresh** nos permite afinar a la hora de elimnar filas o columnas con valores perdidos. Este parámetro indica el número mínimo de valores no perdidos que debe tener la fila o columna para esta sea eliminada.

```python
df.dropna(axis = 'rows', thresh=3)
```

### Rellenando valores perdidos

En determinadas ocasiones es posible que no sea óptimo la eliminación de filas y columnas, es posible que queremos reemplazar estos valores perdidos por algún valor que nosotros consideremos válido. Este valor podría ser por ejemplo cero, o algún tipo de imputación o interpolación que consideremos oportuna. Para esto Pandas dispone del método **fillna()**, que nos retorna una copia del array con los valores reemplazados por el valor que le indiquemos.

```python
#Nos creamos una serie
data = pd.Series([1, np.nan, 3, None], index = list('abcd'))
data
```

En caso de que no exista el valor siguiente al valor perdido, se sigue manteniendo el valor NaN.

Para el caso de los objetos tipo DataFrame el funcionamiento es similar, pero además podemos seleccionar el parámetro **axis** que nos permitirá realizar estas operaciones por filas o columnas.

```python
print(df)
```

```python
df.fillna(method = 'ffill', axis = 'rows')
```

```python
df.fillna(method = 'bfill', axis = 'rows')
```

# Indexación jerárquica

Hasta este punto nos hemos centrados en datos de una o dos dimensiones, almacenados como objetos de tipo Pandas Series y objetos de tipo Pandas DataFrame respectivamente. En determinadas ocasiones es necesaria ir más allá y almacenar datos de una dimensión mayor, es decir, podemos tener los datos indexados por más de una clave. La forma de hace esto, es hacer uso de la multi-indexación que nos permite agregar múltiples índices dentro de un mismo índice. De esta forma los datos de una dimensión alta pueden ser representados de forma compacta dentro de los objetos de una sola dimensión Panda Series o objetos de dos dimensiones Pandas DataFrame.

## Indexación múltiple en Series

Vamos a comenzar a ver como podríamos representar datos en dos dimensiones dentro de una serie de una sola dimensión. Para ser concretos vamos a considerar una serie de datos donde cada elementos tiene un valor de tipo carácter y una clave de tipo numérica. 

### La forma no adecuada

Supongamos que nos gustaría tratar datos de estados de dos años diferentes. Haciendo uso de las herramientas que ya hemos visto en Python, podriamos intentar hacer uso de tuplas como claves.

```python
#Nos creamos los índices
index = [('California', 2000), ('California', 2010), ('New York', 2000), ('New York', 2010), ('Texas', 2000),
        ('Texas', 2010)]

#Nos creamos los valores para dichos índices
populations = [33871648, 37253956, 18976457, 19378102, 20851820, 25145561]

#Nos creamos nuestra Serie
pop = pd.Series(populations, index = index)

#Mostramos el resultado
pop
```

Con este esquema de indexación, podríamos hacer una selección de la siguiente forma

```python
#Hacemos una selección múltiple
pop[('California', 2000): ('California', 2010)]
```

Pero la eficiencia y la comodidad termina aquí. Por ejemplo, si queremos seleccionar todos los datos del año 2010, debemos de realizar algo bastante lento e ineficiente como: 

```python
pop[[i for i in pop.index if i[1] == 2010]]
```

Esto proporciona el resultado esperado, pero tiene dos inconveniente: no es limpio y no es eficiente para conjunto de datos con una gran cantidad de variables.

### La forma correcta: Pandas multi-indexado

Afortunadamente, Pandas nos proporciona una mejor forma de realizar esto. Esencialmente nuestro lista de tuplas, es un elemento multíndice, y el tipo **MultiIndex** de Pandas nos proporciona el tipo de dato deseado.

```python
#Nos creamos un objeto de tipo multi-index
index = pd.MultiIndex.from_tuples(index)
index
```

Podemos ver como ahora nuestro objeto contiene multiples niveles de indexación, en este caso tenemos los estados y los años. Además podemos ver lo valores usados para codificar cada uno de los valores. Si ahora hacemos un reindexado de nuestros datos haciendo uso de **reindex** tenemos el siguiente resultado:

```python
pop = pop.reindex(index)
```

Podemos ver como las dos primeras columnas representan nuestro multi-indexado, mientras que la tercera columna muestra los datos. En el primer nivel de indexación podemos observar que tenemos valores en blanco, esto indica que ese valor toma el valor que tiene justo encima. Ahora por ejemplo si deseamos acceder a los datos del año 2010 podemos hacer uso de una simple indexación en Python:

```python
pop[:, 2010]
```

El resultado podemos ver que se trata de un array individual con las claves de cada uno de los estados y su valor la población que es justo lo que queríamos. Podemos ver que esta forma de extraer los resultados es mucho más limpia y además es mucho más eficiente.

### MultiIndexado como dimensión extra

En el caso anterior podríamos notar algo como lo siguiente: nosotros podríamos almacenar de forma sencilla los datos anteriores en un objeto de tipo Pandas DataFrame con una indexación por filas y columnas. El método **unstack()** convierte un serie con indexación múltiple en un DataFrame convencional.

```python
pop_df = pop.unstack()
print(pop_df)
```

Por otro lado, el método **stack()** hace la operación contraria

```python
pop = pop_df.stack()
print(pop)
```

Tras ver esto nos podríamos preguntar cual es la razón de hacer uso de la multi-indexación jerárquica. La razón es muy simple: de igual forma que pudimos hacer uso de la indexación múltiple para representar datos de dos dimensiones dentro de una Serie de una sola dimensión, podemos hacer uso de la multi indexación para representar datos de tres o más dimensiones en objetos tipo Serie o tipo DataFrame. Cada nivel extra en un multi-indexado representa una dimensión extra de los datos, aprovechando esto obtenemos una mayor flexibilidad en los tipos de datos que podemos representar. Por ejemplo podríamos querer añadir otra columna que nos aporte una mayor información a nuestros datos demográficos, por ejemplo el número de personas menores de 18 años para cada año y para cada uno de los estados, con el MultiIndexado es tan simple como añadir otra columna a nuestro conjunto de datos.

```python
pop2 = pd.DataFrame({'total': pop, 'under18': [9267089, 9284094,4687374, 4318033,5906301, 6879014]})
print(pop2)
```

Todas las funciones universales vistas funcionan también con el multi-indexado jerárquico. Por ejemplo, podemos calcular el porcentaje de gente que tiene menos de 18 años por año en cada uno de los estados.

```python
f_u18 = pop2['under18'] / pop2['total']
f_u18.unstack()
```

El multi-indexado nos permite explorar de una forma sencilla y rápida datos de elevada dimensionalidad.

```python
#Definimos una función que nos permite crear de forma más sencilla un conjunto de datos
def make_df(cols, ind):
    '''Quickly make a DataFrame'''
    data = {c : [str(c) + str(i) for i in ind] for c in cols}
    return pd.DataFrame(data,ind)

#Nos creamos el DataFrame
make_df('ABC', range(3))
```

## Simple concatenación haciendo uso de pd.concat

Pandas dispone de la función **pd.concat()**, que tiene una sintaxis muy similar a **np.concatenate**, pero contiene una serie de argumentos adicionales. **pd.concat()** puede ser usada como una simple concatenación entre objetos de tipo Serie o tipo DataFrame.

```python
#Nos creamos dos series
ser1 = pd.Series(['A', 'B', 'C'], index = [1, 2, 3])
ser2 = pd.Series(['D', 'E', 'F'], index = [4, 5, 6])

#Concatenamos las series
pd.concat([ser1, ser2])
```

Este método también funciona para concatenar objetos de una dimensión mayor como DataFrames.

```python
#Nos creamos dos dataframes
df1 = make_df('AB', [1,2])
df2 = make_df('AB', [3,4])

#Concatenamos 
print(pd.concat([df1,df2]))
```

Por defecto la concatenación se realiza por filas, al igual que **np.concatenate**, la función **pd.concat()** dispone del argumento adicional **axis** que nos permite realizar esta concatenación por columnas.

```python
#Nos creamos dos dataframes
df1 = make_df('AB', [0, 1])
df2 = make_df('CD', [0, 1])

#Concatenamos los dos dataframes por columnas
print(pd.concat([df1, df2], axis = 1))
```

### Índices duplicados

Una diferencia entre **np.concatenate** y **pd.concat()** es que la función **pd.concat()** preserva los índices, incluso si el resultado tiene índices duplicados.

```python
#Nos creamos dos dataframes
df1 = make_df('AB', [0, 1])
df2 = make_df('AB', [2, 3])

#Duplicamos índices
df1.index = df2.index

#Concatenamos
pd.concat([df1, df2], axis = 0)
```

Podemos notar que tenemos índice repetidos en el resultado. Aunque este resultado es válido para DataFrames el resultado no es deseable, para solucionar este problema dispone de varias formas

#### Capturando las repeteciones como un error

Si lo queremos es comprobar que los índices tras realizar la operación de concatenación no están repetidos, podemos especificar el flag **verify_integrity**. Si activamos este flag, dándole el valor de **True** generará una excepción en caso de que existan índices duplicados.

```python
try:
    pd.concat([df1, df2], verify_integrity=True)
except ValueError as e:
    print('Value error:', e)
```

#### Ignorando los índices

Existen situaciones en las cuales los índices no son importantes, y simplemente podemos ignorarlos. Para especificar esto disponemos del flag **ignore_index**. Si le damos el valor **True**, la concatenación se creará con un nuevo índice de enteros.

```python
#Concatenamos haciendo uso del parámetro ignore_index
pd.concat([df1, df2], ignore_index=True)
```

#### Agregando claves multi-índice

Otra opción es hacer uso del parámetro **keys** para especificar una etiqueta a nuestra fuente de datos, el resultado será una indexación jerárquica.

```python
#Concatenamos haciendo uso del parámetro keys
pd.concat([df1,df2], keys = ['df1', 'df2'])
```

## Concatenación mediante joins

En los ejemplos que hemos visto anteriormente hemos realizado la concatenación de dataframes que compartian columnas. En la práctica, los datos que vienen de fuentes diferentes podrían tener diferentes columnas, en estas situacios la función **pd.concat()** ofrece distintas opciones. Consideremos la concatenación de dos DataFrames, que comparten algunas columnas y otras no.

```python
#Nos creamos dos dataframes
df1 = make_df('ABC', [1,2])
df2 = make_df('BCD', [3,4])

#Concatenamos 
pd.concat([df1, df2])
```

Por defecto los datos que no machean son rellenados por NaN. Para cambiar esto, podemos especificar los parámetros **join** y **join_axes**. Por defecto, el join es una unión de las columnas de entrada (join = outer), pero podemos cambiar esto a una intersección de columnas (join = inner).

```python
#Hacemos un join inner
pd.concat([df1, df2], join = 'inner')
```

Otra opción es especificar los índices por los cuales queremos realizar nuestro join, para ellos disponemos del parámetro **join_axes**, el cual toma una lista de objetos tipo índice.

```python
#Hacemos el concat haciendo uso de join_axes
pd.concat([df1, df2], join_axes = [df1.columns])
```

# Combinando datasets: Merge y Join

Una característica esencial de Pandas es que ofrece un alto rendimiento en memoria en las operaciones de **merge** y **join**.

## Algebra relacional

El comportamiento de **pd.merge()** está basado en el algebra relacional, se trata de un conjunto de reglas para la manipulación de datos relacionales. La fuerza del enfoque del álgebra relacional es que propone varias operaciones primitivas, que se convierten en bloques de operaciones más complejas para un conjunto de datos.

## Categorías de joins

La función **pd.merge()** implementa varios tipos de joins: el **one-to-one**, **many-to-one** y **many-to-many**. Los tres tipos de join son accesibles de forma idéntica a través de la función **pd.merge()**, el tipo de join realizado depende la forma de los datos de entrada. 

### One-to-one joins

Se trata del tipo de join más simple, el cual en muchas ocasiones es muy similar a la concatenación por columna. Consideremos los siguientes conjuntos de datos.

```python
#Nos creamos los dos dataframes
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})

df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
'hire_date': [2004, 2008, 2012, 2014]})
```

La combinación de estos dos conjuntos de datos en uno solo puede hacerse haciendo uso de la función **pd.merge()**.

```python
#Hacemos el merge entre los dos cojuntos de datos
df3 = pd.merge(df1, df2)
print(df3)
```

La función **pd.merge()** es capaz de reconocer que ambos conjuntos de datos comparten la columna **employee**, y de forma automática realiza la unión por esta columna, es decir, usa dicha columna como clave. El resultado de la unión es un nuevo DataFrame que contiene la info de las dos entradas. Además la función **pd.merge()** no mantiene el índice, salvo que esta unión se haga por el índice.

### Many-to-one joins

Se tratan de joins en los cuales una de las dos columnas clave contiene entradas duplicadas. En este caso el DataFrame resultante mantendrá estos valores duplicados. 

```python
#Nos creamos un nuevo DataFrame
df4 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'],
'supervisor': ['Carly', 'Guido', 'Steve']})

print(df4)
```

```python
#Hacemos el merge
print(pd.merge(df3, df4))
```

### Many-to-many joins

Este tipo de joins son bastante confusos conceptualmente, pero sin embargo están bien definidos. Si las dos columnas clave, contienen valores duplicados, entonces el resultado es un many-to-many merge. Supongamos el siguiente conjunto de datos.

```python
df5 = pd.DataFrame({'group': ['Accounting', 'Accounting','Engineering', 'Engineering', 'HR', 'HR'],
                    'skills': ['math', 'spreadsheets', 'coding', 'linux','spreadsheets', 'organization']})
```

```python
print(df1); print(df5);
```

```python
#Hacemos el merge
print(pd.merge(df1,df5))
```

En la práctica esta forma de hacer la unión entre conjunto de datos no es adecuado ni limpia, a continuación veremos una forma más adecuada, que es hacer el join mediante merge pero especificando la clave.

## Especificando la clave para el Merge

Ya hemos visto el uso del **merge** de forma básica: su objetivo es encontrar columnas que sean comunes entre ambos conjuntos de datos y usar esta como clave para realizar la unión. Sin embargo, es posible que el nombre de las columnas que deben de hacer como clave  no coincidan, para estas situaciones **merge** dispone de varias opciones.

### Parámetro on

Es muy simple, podemos especificar de forma explícita el nombre de la columna que queremos que sea usada como clave, este valor puede tomar el nombre de la columna o bien una lista de columnas.

```python
print(df1);print(df2)
```

```python
#Hacemos uso del merge
print(pd.merge(df1, df2, on = 'employee'))
```

### Los argumentes left_on y right_on

En determinadas situaciones podemos querer unir dos conjuntos que tienen diferente nombre en sus columnas. En este caso podemos hacer uso de los argumentos **left_on** y **rigth_on**.

```python
#Nos creamos nuestro conjunto de datos
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
'salary': [70000, 80000, 120000, 90000]})
```

```python
#Mostramos los conjuntos de datos que deseamos unir
print(df1);print(df3)
```

```python
#Hacemos la unión 
print(pd.merge(df1, df3, left_on = 'employee', right_on = 'name'))
```

Podemos ver que el resultado contiene una columna redundante. Podemos eliminar esta columna haciendo uso del método **drop()**.

```python
pd.merge(df1, df3, left_on = 'employee', right_on = 'name').drop('name', axis = 1)
```

### Los argumentos left_index and right_index

Existen situaciones en las que, en lugar de hacer una unión a través de columnas deseamos realizarlas a través de índices. Supongamos la siguiente situación.

```python
#Asignamos la columna employee para df1 y df2 como índice
df1_ind = df1.set_index('employee')
df2_ind = df2.set_index('employee')
```

```python
#Vemos el resultado
print(df1_ind);print(df2_ind)
```

Podemos especificar mediante los argumentos **left_index** y **right_index** que queremos hacer la unión por dichos índices.

```python
#Hacemos el merge
print(pd.merge(df1_ind, df2_ind, left_index = True, right_index = True))
```

Por conveniencia, los objetos tipo DataFrame disponen del método **join()**, el cual realiza la unión por defecto mediante índices.

```python
#Hacemos uso del join
print(df1_ind.join(df2_ind))
```

Si lo deseamos podemos hacer un mix a la hora de unir conjuntos de datos mezclando columnas y índices, es decir, podemos combinar **left_index** con **right_on** o **left_on** con **right_index**.

```python
print(df1_ind); print(df3)
```

## Especificando operaciones aritméticas en joins

Es posible que existan situaciones en las cuales uno de los valores de nuestra columna clave por la cual realizamos la unión aparezca en una de las columnas pero no en la otra.

```python
#Supongamos el siguiente conjunto de datos
df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
'food': ['fish', 'beans', 'bread']},
columns=['name', 'food'])

df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],
'drink': ['wine', 'beer']},
columns=['name', 'drink'])

#Mostramos ambos conjuntos 
print(df6)
```

```python
print(df7)
```

Podemos ver que si hacemos la unión por la columna clave **name** tenemos valores que aparecen en un conjunto de datos y en el otro no.

```python
#Realizamos la unión
print(pd.merge(df6, df7))
```

Podemos que el resultado es tan solo una fila, que se trata del valor que aparace en ambos conjuntos de datos. Esto es debido a que por defecte la función merge realiza lo que se conoce como **inner join**, es decir, se realiza la intersección entre ambos conjuntos de datos. Podemos especificar el tipo mediante el parámetro **how**.

```python
print(pd.merge(df6, df7, how = 'inner'))
```

Otros valores que puede tomar el argumento **how** son: **outer**, **left**, **right**. Dándole el valor **outer** retorna un conjunto de datos que es la unión de ambos conjuntos de datos, los valores perdidos son rellenados como **NaN**.

```python
#Hacemos la unión con outer
print(pd.merge(df6,df7, how = 'outer'))
```

Los valores **left_join** y **right_join** nos retornan las uniones respecto a los dataframes respectivos. Es decir, si hacemos un left_join mantendremos los valores de la tabla de la izquierda más la unión de estos valores con la tabla de la derecha.

```python
#Hacemos uso de left_join
print(pd.merge(df6, df7, how = 'left'))
```

```python
#Hacemos uso de righ_join
print(pd.merge(df6, df7, how = 'right'))
```

## Groupby: Split, Apply, Combine

Si bien las agregaciones simples pueden darle un significado a nuestros datos, en determinadas situaciones es posible que queramos realizar estas agregaciones a partir de series índices o etiquetas: esto puede ser realizado a partir de la operación **groupby**.

### Split, Apply, Combine

* El paso split implica dividir y agrupar un DataFrame, según el valor de la clave especificada.

* El paso apply implica aplicar una función usualmente de agregación, transformación o filtrado a cada uno de los grupos individuales.

* El paso combine finalmente unifica los resultados en un solo array o dataframe.

La potencia de la operación **groupby** es que permite al usuario ignorar como se realizan todas estas operaciones por detrás.

```python
#Nos creamos un conjunto de datos
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns=['key', 'data'])

#Vemos el resultado
df
```

Podemos calcular el conjunto de operaciones split, apply y combine simplemente haciendo uso de la función **groupby** pasándole la clave por la cual deseamos realizar esto.

```python
df.groupby('key')
```

Podemos notar que no retorna un objeto de tipo DataFrame, sino que retorna un objeto de tipo DataFrameGroupBy. Esto objeto, se trata de un objeto que este preparado para atacar a los grupos , pero no realiza un cálculo real hasta que no se realiza algún tipo de agregación. 

```python
#Aplicamos la agregación de sumatorio
df.groupby('key').sum()
```

### El objeto GroupBy

El objeto de tipo GroupBy se trata de un objeto muy flexible que nos permite realizar una serie de operaciones bastante complejas de una forma más sencilla.

* Los objetos de tipo GroupBy al igual que los objetos de tipo DataFrame permiten la **indexación por columnas**.

```python
#Agrupamos por la columna method
planets.groupby('method')
```

```python
#Seleccionamos solo el periodo orbital de cada agrupacion
planets.groupby('method')['orbital_period']
```

```python
#Para cada agrupación calculamos la media del periodo orbital
planets.groupby('method')['orbital_period'].mean()
```

* Permiten la iteración a través de grupos

```python
for (method, group) in planets.groupby('method'):
    print("{0:30s} shape={1}".format(method, group.shape))
```

Dado un objeto de tipo **Groupby** podemos hacer uso de los distintos métodos vistos para cada uno de los grupos por ejemplo podemos aplicar el método **describe()**.

```python
planets.groupby('method')['year'].describe()
```

Aquí podemos ver que la mayoría de exoplanetas descubiertos han sido por los métodos RadialVelocity y Transit.

### Aggregate, filter, transform, apply

En el apartado anterior nos hemos centrado en la agregación para la operación de combinación, sin embargo existen muchas otras posibilidades. En particular, los objetos **GroupBy()** tienen **aggregate()**, **filter()**, **transform()** y **apply()** que implementan de una forma muy eficiente una gran variedad de operaciones realmente muy utiles.

```python
#Nos creamos un conjunto de datos
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
'data1': range(6),
'data2': rng.randint(0, 10, 6)},
columns = ['key', 'data1', 'data2'])

#Vemos el resultado 
df
```

* **Agregación**

Hasta ahora estamos famicializados con las agregaciones **sum()**, **mean()** etc para la función **GroupBy**. La función **aggregate()** nos permite una mayor flexibilidad, esta funció puede tomar una lista de agregacioes y computar todas de una sola vez. 

```python
#Hacemos uso de aggregate
df.groupby('key').aggregate(['min', np.median, max])
```

Otra opción es hacer uso de un diccionario y mapear de esta forma que agregación queremos aplicar a cada una de las columnas.

```python
df.groupby('key').aggregate({'data1': 'min',
                            'data2': 'max'})
```

* **Filtrando**

Esta operación nos permite eliminar datos en función de propiedades del grupo. Por ejemplo nos podría interesar mantener solo aquellos grupos en los que su desviación estándar es mayor a un cierto valor crítico.

```python
#Nos creamos la función que realiza el filtrado
def filter_func(x):
    return x['data2'].std() > 4

#Aplicamos la función filter
df.groupby('key').filter(filter_func)
```

La función **filter()** debería retornar un valor booleano que especifica si el grupo pasa o no el filtro. Debido a que el grupo A no tiene un desviación estándar superior a 4, este grupo es eliminado del resultado.

* **Transformación**

Mientras que la agregación retorna una versión reducida de los datos, la transformación retorna una versión de los datos transformada pero completa. Por lo tanto la salida tras aplicar una transformación es de igual tamaño que los datos de entrada.

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

* **El método Apply()**

Este método nos permite aplicar una función arbitraria a cada uno de los grupos. La función debería tomar un DataFrame y devolver un objeto tipo Pandas (DataFrame o Series) o escalar, la operación combinación se adaptará al tipo de dato devuelto.

```python
#Creamos una función que nos permite normalizar la primera columna como la suma de la segunda
def norm_by_data2(x):
    x['data1'] /= x['data2'].sum()
    return x

#Aplicamos la función con apply
df.groupby('key').apply(norm_by_data2)
```

```python
df.groupby(df['key']).sum()
```

# Pivot Tables

Ya hemos visto como la función **GroupBy** nos permite obtener relaciones dentro de nuestro conjunto de datos. **Pivot Table** es una operación similar que se ve mucho en las hojas de cálculo y otros programas con datos tabulares. Una **pivot table** toma como entrada datos de columnas simples, y agrupa las entradas en una tabla de dos dimensiones, de forma que nos proporciona una sumarización de los datos multi-dimensional. La diferencia entre la operación **GroupBy** y las **pivot tables** en ocasiones puede causar confusión, podemos pensar en las **pivot tables** como una versión multi-dimensional de la función **GroupBy**. Cuando aplicamos **split-apply-combine**, tanto la operación **split** como el **combine** pasan a través no solo de un índice uni-dimensional, sino a través de un grid de dos dimensiones.

## Motivación de las Pivot Tables

A continuación vamos hacer uso del conjunto de datos Titanic.

```python
import seaborn as sns
import pandas as pd
import numpy as np
titanic = sns.load_dataset('titanic')
titanic.head(6)
```

## Pivot tables a mano

Para comprender mejor este tipo de datos, vamos a comenzar agrupando de acuerde con el sexo, supervivencia, y combinaremos esto de alguna forma. Visto la sección anterior, nos podríamos ver tentados a hacer uso de la funcion **GroupBy** de la siguiente forma.

```python
titanic.groupby('sex')['survived'].mean()
```

Esta información es útil pero podríamos estar interesado en conocer exactamente esta información pero por clase social. 

```python
titanic.groupby(['sex', 'class'])['survived'].mean().unstack()
```

## Sintaxis de las Pivot Tables

A continuación mostramos la sintáxis de una **pivot table** para realizar la operación anterior.

```python
titanic.pivot_table('survived', index = 'sex', columns = 'class')
```

### Pivot tables multi nivel

Al igua que la función **GroupBy** podemos pivotar una tabla a partir de múltiples niveles. Supongamos que estamos interesados en meter la edad como una tercera dimensión.

```python
#Agrupamos la edad en rangos
age = pd.cut(titanic['age'], [0, 18, 80])

#Hacemos uso del pivot table
titanic.pivot_table('survived', index = ['sex', age], columns = 'class')
```

Esta misma estrategia puede ser aplicada a nivel de columna, supongamos que deseamos conocer la cantidad de dinero que se pago por genero, rango de edad y clase. La función **qcut** nos permite calcular los cuantiles.

```python
fare = pd.qcut(titanic['fare'], 2)

titanic.pivot_table('survived', index = ['sex', age], columns = [fare, 'class'])
```

### Opciones adicionales para las pivot tables

Además de los tres argumentos vistos, la función **pivot_table** cuenta con una serie de argumentos adicionales que pueden ser de gran utilidad. Tenemos dos argumentos relacionados con los valores perdidos:

* **dropna**: puede tomar el valor de **True** o **False**. Por defecto toma el valor de **True** elimina aquellas columnas en las que todos sus valores sean NaN.

* **fill_value**: nos permite indicar por que valor queremos rellenar los valores perdidos.

El argumento **aggfunc** nos permite controlar el tipo de agregación que queremos realizar, por defecto toma el valor de la media (**mean**). 

```python
titanic.pivot_table(index = 'sex', columns = 'class', aggfunc = {'survived': sum, 'fare': 'median'})
```

En determinadas ocasiones podemos estar interesados en calcular el total, para esto contamos con el argumento **margins** que puede tomar el valor de **True** o **False**. Si toma el valor de **True** nos calculo el total, por tipo de indexación, por defecto toma el valor de **False**.

```python
titanic.pivot_table('survived', index = 'sex', columns = 'class', margins = True)
```

## Introducción a las operaciones con Strings en Pandas

Ya vimos como herramientas como **Numpy** o **Pandas** generalizas las operaciones aritméticas por lo que es sencillo y rápido y rápido realizar algunas operaciones aritméticas sobre arrays de gran dimensión.

### Remuestreo

Una de las necesidades más comunes a la hora de trabajar con series temporales es hacer un remuestreo a una menor o mayor frecuencia. Podemos hacer esto haciendo uso del método **resample()**, o aún más simple haciendo uso del método **asfreq()**. La principal diferencia entre estos dos métodos es que **resample()** es una agregación de datos mientras que **asfreq()** se trata de una selección de datos.

```python
#Vamos a remuestrear de forma mensual 
f['close_price'].astype('float').plot(alpha = 0.5, style = '-')
f['close_price'].astype('float').resample('BM').mean().plot(style = ':')
f['close_price'].astype('float').asfreq('BM').plot(style = '--')
plt.legend(['input', 'resample', 'asfreq'],loc='upper left');
```

Podemos ver como **resample** reporta para cada punto el promedio del mes anterior, mientras que **asfreq** reporta el valor de fin de mes.

El método **asfreq()** nos permite indicar la forma en la cual queremos imputar los valores perdidos.

```python
fig, ax = plt.subplots(2, sharex=True)
f['close_price'].loc['2018-04'].astype('float').asfreq('D').plot(ax=ax[0], marker='o')
f['close_price'].loc['2018-04'].astype('float').asfreq('D', method='bfill').plot(ax=ax[1], style='-o')
f['close_price'].loc['2018-04'].astype('float').asfreq('D', method='ffill').plot(ax=ax[1], style='--o')
plt.show()
```

### Time-shifts

Otra operación muy común con series temporales es el desplazamiento de los datos en el tiempo. Pandas dispone de dos métodos para realizar esto: **shift()** y **tshift()**. La principal diferencia entre estos dos métodos es que el método **shift()** realiza el desplazamiento de los datos, mientras que **tshift()** realiza el desplazamiento de los índices.

```python
fig, ax = plt.subplots(3, sharey=True)
f_aux = f['close_price'].astype('float').asfreq('D', method = 'pad')
f_aux.plot(ax=ax[0])
f_aux.shift(100).plot(ax=ax[1])
f_aux.tshift(100).plot(ax=ax[2])

local_max = pd.to_datetime('2018-01-12')
offset = pd.Timedelta(100, 'D')

ax[0].legend(['input'], loc=2)
ax[0].get_xticklabels()[4].set(weight='heavy', color='red')
ax[0].axvline(local_max, alpha=0.3, color='red')

ax[1].legend(['shift(900)'], loc=2)
ax[1].get_xticklabels()[4].set(weight='heavy', color='red')
ax[1].axvline(local_max + offset, alpha=0.3, color='red')

ax[2].legend(['tshift(900)'], loc=2)
ax[2].get_xticklabels()[1].set(weight='heavy', color='red')
ax[2].axvline(local_max + offset, alpha=0.3, color='red')

plt.subplots_adjust(hspace=0.5)
plt.show()
```

# Alto rendimiento en Pandas: eval() y query()

## Motivación query() y eval(): expresiones compuestas

Ya hemos visto que tanto Pandas como Numpy soportan operaciones vectorizadas, por ejemplo podemos sumar los elementos de dos arrays.

```python
rng = np.random.RandomState(42)
x = rng.rand(10^6)
y = rng.rand(10^6)
%timeit x + y 
```

Pero esto se convierte menos eficiente cuando estamos haciendo uso de expresiones por ejemplo suponer la siguiente expresión 

```python
mask = (x > 0.5) & (y < 0.5)

#Lo anterior se traduce en 
aux1 = (x > 0.5)
aux2 = (y < 0.5)
mask = aux1 & aux2
```

En otras palabras, cada uno de los parsos intermedios se asigna a memoria. Si x e y se tratan de arrays de una gran dimensionalidad, esto puede dar lugar a coste computacional muy elevado. La librería **Numexpr** nos da la posibilidad de calcular este tipo de expresiones element a elemento, sin necesidad de almacenar arrays intermedias. Se puede consultar la documentación [aqui](https://github.com/pydata/numexpr).

```python
import numexpr
mask_numexpr = numexpr.evaluate('(x > 0.5) & (y < 0.5)')
np.allclose(mask, mask_numexpr)
```

Como ya hemos indicado el beneficio de **Numexpr** evalua las expresiones de forma que no usa matrices temporales, y por lo tanto puede ser mucho más eficiente que Numpy especialmente con arrays de gran tamaño. Las herramientas **eval()** y **query()** que veremos funcionan de forma similar y dependen de **Numexpr**.

## pandas.eval() para operaciones eficientes

La función **eval()** de Pandas hace uso de expresiones de tipo string para realizar operaciones de una forma más eficiente usando DataFrames. Por ejemplo consideremos los siguientes DataFRames:

```python
#Nos creamos los conjuntos de datos
nrows, ncols = 100000,100
rng = np.random.RandomState(42)
df1, df2, df3, df4 = (pd.DataFrame(rng.rand(nrows, ncols)) for i in range(4))
```

```python
%timeit df1 + df2 + df3 + df4
```

```python
%timeit pd.eval('df1 + df2 + df3 + df4')
```

La función **eval()** es un 50% más rápida y además hace uso de una menor cantidad de memoria.

### Operaciones soportadas por eval()

La función **eval()** soporta un amplio rango de operaciones. Supongamos los siguientes conjuntos de datos

```python
df1, df2, df3, df4, df5 = (pd.DataFrame(rng.randint(0, 1000, (100,3))) for i in range(5))
```

#### Operadores aritméticos

```python
result1 = -df1 * df2 / (df3 + df4) - df5
result2 = pd.eval('-df1 * df2 / (df3 + df4) - df5')
np.allclose(result1, result2)
```

#### Operadores de comparación

```python
result1 = (df1 < df2) & (df2 <= df3) & (df3 != df4)
result2 = pd.eval('(df1 < df2) & (df2 <= df3) & (df3 != df4)')
np.allclose(result1, result2)
```

#### Operadores bit a bit

```python
result1 = (df1 < 0.5) & (df2 < 0.5) | (df3 < df4)
result2 = pd.eval('(df1 < 0.5) & (df2 < 0.5) | (df3 < df4)')
print(np.allclose(result1, result2))

#También acepta los operadores and y or
result3 = pd.eval('(df1 < 0.5) and (df2 < 0.5) or (df3 < df4)')
print(np.allclose(result3, result1))
```

#### Objetos atributos e índices

```python
result1 = df2.T[0] + df3.iloc[1]
result2 = pd.eval('df2.T[0] + df3.iloc[1]')
np.allclose(result1, result2)
```

Otras operaciones como: llamadas a funciones, condicionales (if), bucles etc no están implementados en **eval()**. Si queremos ejecutar este tipo de expresiones más complicadas podemos hacer uso de **Numexpr**.

### DataFrame.eval() para operaciones en columnas

Así como pandas dispone de una función **eval()**, los objetos de tipo DataFrame tienen un método **eval()** que trabaja de forma similar. El beneficio del método **eval()** es que las columnas pueden ser referenciadas a través del nombre. Supongamos el siguiente DataFrame:

```python
df = pd.DataFrame(rng.rand(1000, 3), columns = ['A', 'B', 'C'])
df.head()
```

Podríamos hacer uso de pd.eval para hacer operaciones entre columnas

```python
result1 = (df['A'] + df['B']) / (df['C'] - 1)
result2 = pd.eval("(df.A + df.B) / (df.C - 1)")
np.allclose(result1, result2)
```

La ventaja de df.eval es que nos permite operar directamente con el nombre de las columnas

```python
result3 = df.eval('(A + B) / (C - 1)')
np.allclose(result1, result3)
```

Además de todo los visto mediante **df.eval()** también podemos crear columnas

```python
df.eval('D = (A + B) / C', inplace = True)
df.head()
```

También podemos modificar columnas ya existentes

```python
df.eval('D = (A-B)/C', inplace = True)
df.head()
```

**df.eval()** también soporta una sintáxis adicional que permite trabajar con variables locales en Python. 

```python
#Obtenemos la media de las columnas
column_mean = df.mean(axis = 1)
result1 = df['A'] + column_mean
result2 = df.eval('A + @column_mean')
np.allclose(result1, result2)
```

El carácter **@** hace referencia a una variable en lugar de una columna, esto nos permite operar de forma eficiente este tipo de expresiones.

## Método query para DataFrames

Los objetos de tipo DataFrame tienen otro tipo de método basado en la evaluación de strings, llamado **query()**. Consideremos el siguiente ejemplo:

```python
result1 = df[(df.A < 0.5) & (df.B < 0.5)]
result2 = pd.eval('df[(df.A < 0.5) & (df.B < 0.5)]')
np.allclose(result1, result2)
```

El método query se trata de un método que es mucho más eficiente que las máscaras y además es mucha más sencillo a la hora de comprender.

```python
result3 = df.query('A < 0.5 & B < 0.5')
np.allclose(result1, result3)
```

```python
Cmean = df['C'].mean()
result1 = df[(df.A < Cmean) & (df.B < Cmean)]
result2 = df.query('A < @Cmean and B < @Cmean')
np.allclose(result1, result2)
```

## Cuando hacer uso de estas funciones

A la hora de hacer uso de estas funciones, es debido a dos consideraciones: tiempo de computo y uso de memoria.El uso  de memoria es el aspecto más predecible. Cada calculo realizado mediante Numpy o mediante Pandas implica de forma implícita la creación de arrays temporales. Si el tamaño de nuestras arrays o DataFrames son considerables comparándolo con el sistema de almacenamiento de nuestra computadora, entonces es una buena idea hacer uso de las funciones **eval()** y **query()**. Podemos chequear el tamaño de nuestro array de forma aproximada mediante:

```python
df.values.nbytes
```

**eval()** puede ser más rápido incluso si no estamos intentando maximizar la memoria de nuestra computadora. Cuando creamos arrays temporales lo que hacemos es comparar estos arrays con el caché de nuestra CPU, eval() al no crear estos DataFrames temporales evita el movimiento potencialmente lento de los valores entre las diferentes cachés. La experiencia demuestra que el tiempo computacional no es relativamente significativo entre los métodos tradicionales y entre **eval()** y **query()**, incluso los métodos tradicionales en determinadas ocasiones son más rápidos para arrays pequeñas, el principal motiva a la hora de hacer uso de **eval()** y **query()** es el ahorro de memoria.[Aqui](http://pandas-docs.github.io/pandas-docs-travis/) podemos encontrar toda la documentación en lo que respecta a Pandas
